<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

```r
frecuencia_abs_acum <- function(dt) { # Input class <- vector
    frecuencia_abs <- table(dt) # Obtenemos la frecuencia absoluta de los elementos del vector
    for(index in 1:length(frecuencia_abs)) { # Recorremos todos los elementos del vector para realizar una suma progresiva
        if(index != 1) { # El primer elemento se queda como está
            frecuencia_abs[index] <- frecuencia_abs[index -1] + frecuencia_abs[index]
        } # Se realiza la suma teniendo en cuenta el valor anterior
    }
    return(frecuencia_abs)
}
```

```r
frecuencia_rel_acum <- function(dt) { # Input class <- vector
    frecuencia_abs <- prop.table(table(dt)) # Obtenemos la frecuencia relativa de los elementos del vector
    for(index in 1:length(frecuencia_abs)) { # Recorremos todos los elementos del vector para realizar una suma progresiva
        if(index != 1) { # El primer elemento se queda como está
            frecuencia_abs[index] <- frecuencia_abs[index -1] + frecuencia_abs[index]
        } # Se realiza la suma teniendo en cuenta el valor anterior
    }
    return(frecuencia_abs)
}
```

```r
frecuencia_rel_acum(flores)
```

<div style="background-color:#5DADE2"><b>Apartado 7</b></div>

```r
frecuencia_abs <- function(dt) { # Input class <- vector
    valores_contados <- c() # Tenemos que hacer una leyenda donde veamos todos los valores evaluados
    total_contados <- c() # Tenemos una leyenda que acompaña a los valores y contiene a sus totales dentro del vector
    for(value in dt) { # Primer for evalua para todos los elementos
        contador <- 0 # Variable de control de valores en el vector
        if(!(value %in% valores_contados)) {
            for(values in dt) { # Segundo for evalua el vector en base al valor pillado
                if(value == values) {
                    contador <- contador + 1 # Como ha conseguido match, suma 1 a la cuenta
                }
            }
        }
        total_contados <- c(total_contados, contador) # Añadimos al vector de los nombres, el nombre utilizado
        valores_contados <- c(valores_contados, value) # Añadimos la cuenta total que nos ha salido al vector de totales
    }
    names(total_contados) <- valores_contados # Asignamos nombres al vector
    return(total_contados)
}
```

```r
frecuencia_abs(flores) # Debería de salir algo parecido
```

```r
frecuencia_rel <- function(dt) { # Input class <- vector
    valores_contados <- c() # Tenemos que hacer una leyenda donde veamos todos los valores evaluados
    total_contados <- c() # Tenemos una leyenda que acompaña a los valores y contiene a sus totales dentro del vector
    total <- length(dt) # Obtenemos el total de valores que hay dentro del vector
    for(value in dt) { # Primer for evalua para todos los elementos
        contador <- 0 # Variable de control de valores en el vector
        if(!(value %in% valores_contados)) {
            for(values in dt) { # Segundo for evalua el vector en base al valor pillado
                if(value == values) {
                    contador <- contador + 1 # Como ha conseguido match, suma 1 a la cuenta
                }
            }
        }
        contador <- contador / total # Sacamos el porcentaje en base a 1%
        total_contados <- c(total_contados, contador) # Añadimos al vector de los nombres, el nombre utilizado
        valores_contados <- c(valores_contados, value) # Añadimos la cuenta total que nos ha salido al vector de totales
    }
    names(total_contados) <- valores_contados # Asignamos nombres al vector
    return(total_contados)
}
```

```r
frecuencia_rel(flores) # Algo parecido
```

Como vamos a tener que buscar varios elementos en varios apartados, vamos a realizar una función que tome como input los valores del rango que buscamos, primero el mínimo, después el máximo. De ahí que devuelva el conteo final

```r
conteo_en_rango <- function(minimo, maximo, dt) { # Declaración de la función
    conteo <- 0 # Declaración del conteo de datos
    for(value in dt) {
        if(value >= minimo & value <= maximo) {
            conteo <- conteo + 1 # Si está dentro del rango, se cuenta
        }
    }
    porcentaje <- conteo / length(dt) # Devolvemos el número literal y el porcentaje que representa en el vector
    output <- c(conteo, porcentaje)
    names(output) <- c("Nº", "%")
    return(output)
}
```

```r
conteo_en_rango(35, 39.9, ventas)
```

<div style="background-color:#5DADE2"><b>Apartado 5</b></div>

```r
prueba1 <- conteo_en_rango(28, 28.9, ventas)
prueba2 <- conteo_en_rango(29, 29.9, ventas)
```

```r
prueba1 < prueba2 # El rango de entre 28-28.9 es mayor que el de el 29-29.9 mil
```

```r
varianza = function(x) {
    media = mean(x)
    suma_cuadrados = 0
    for(i in x){
        suma_cuadrados = suma_cuadrados + (i - media) * (i - media)
    }
    return (suma_cuadrados / length(x))
}
```

```r
varianza(edad)
```

```r
var(edad)  # la diferencia entre los resultados de la funcion varianza y var es debido a la forma de aproximar los decimales de cada una.
```
