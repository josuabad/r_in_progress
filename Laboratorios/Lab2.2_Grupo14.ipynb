### Ejercicio 1

Supongamos que gestionamos una pequea empresa que vende un producto online. Basndonos en histricos, sabemos que, en promedio, el 20% de las personas que visitan nuestro sitio web realizan una compra. Ahora, en un da particular, tenemos 500 visitantes en el sitio. Asumir que la variable aleatoria $x$ (nmero de compras) sigue una distribucin binomial, ya que tenemos un nmero fijo de ensayos (500 visitantes), cada ensayo es independiente, y cada ensayo resulta en xito (compra) con una probabilidad del 20%.

1. Qu distribucin de probabilidad tiene $x$? Tenis que justificar el tipo de distribucin y sus parmetros.


2. Cul es la probabilidad de tener exactamente 100 compras en un da, dado que 500 personas visitan el sitio?


3. Cul es la probabilidad de tener ms de 100 compras en un da, dado que 500 personas visitan el sitio?


4. Cul es la probabilidad de tener menos de 100 compras en un da, dado que 500 personas visitan el sitio?


5. Representar grficamente la funcin de masa de la variable aleatoria $x$ e interpretar los resultados.


6. Representar grficamente la funcin de distribucin de la variable aleatoria $x$ e interpretar los resultados.

<div style="background-color:#F7DC6F"><b>Resolucin Ejercicio 01</b></div>

<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

La distribucin de probabilidad de X puede ser representada por una funcin de masa binomial porque podemos calcular exactamente la cantidad de personas que han comprado, lo cual representa un xito para nosotros

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

```r
dbinom(x = 100, size = 500, prob = 0.2)
```

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

```r
1 - pbinom(q = 100, size = 500, prob = 0.2)
```

Realizamos 1 - la distribucin acumulada porque con la acumulada calculamos que sea 100 o menos y justo queremos lo contrario. Entonces, al estar dispuesta en tanto por 1, le restamos 1.

<div style="background-color:#5DADE2"><b>Apartado 4</b></div>

```r
pbinom(q = 100, size = 500, prob = 0.2) - dbinom(x = 100, size = 500, prob = 0.2)
```

Como tiene que ser menos de 100, y la acumulada lo calcula a partir de 100 inclusive, le restamos a la acumulada la probabilidad de tener justo 100

<div style="background-color:#5DADE2"><b>Apartado 5</b></div>

```r
set.seed(1)
func_masa <- dbinom(0:500, size = 500, prob = 0.2)
```

```r
plot(func_masa, type = "l")
```

La grfica nos ensea la probabilidad de que todos los clientes compren de forma separa, es decir, que compren un nmero de clientes concretos, podran ser 1, 2, 3, 4, 5, 30, 58, 100, ...
<br>
Con esto nos damos cuenta que si realizamos la grfica de las probabilidades acumuladas, a partir del valor 100 hay un pico de incremento

<div style="background-color:#5DADE2"><b>Apartado 6</b></div>

```r
set.seed(1)
func_acum <- pbinom(q = 0:500, size = 500, prob = 0.2)
```

```r
plot(func_acum, type = "l")
```

Nos sale justo lo comentado en el apartado anterior. Otra cosa que debemos comentar es que ms o menos a partir del valor 130 la probabilidad de que compren 130 clientes en un da es muy bajita, por eso la grfica se achata y se convierte en horizontal

4. Generar una muestra de tamao $1000$ de una distribucin geomtrica siendo $p=0.05$ y luego calcular las frecuencias muestrales y compararlas con las probabilidades tericas.

<div style="background-color:#F7DC6F"><b>Resolucin Ejercicio 02</b></div>

<div style="background-color:#5DADE2"><b>Apartado 1</b></div>

```r
# Tenemos que usar la distribucion de probabilidad geometrica ya que estariamos realizando repeticiones independientes
# de un experimento que tiene una respueta binaria y la probabilidad de que salga defectuosa la pieza es siempre la misma
```

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

```r
dgeom(10,0.05)
```

<div style="background-color:#5DADE2"><b>Apartado 3</b></div>

```r
1 - pgeom(4,0.05) # P(x > 4) Probabilidad de que no salga un fallo en los primeros 4 intentos
dgeom(4,0.05) # P(x = 4) Probabilidad de que salga un fallo en el cuarto intento (y que no salga en los primeros 3)
pgeom(4,0.05) # P(x <= 4) Probabilidad de que salga un fallo en alguno de los primeros 4 intentos
pgeom (4,0.05) - dgeom (4,0.05) # P(x < 4) Probabilidad de que salga un fallo en alguno de los 3 primeros intentos
```

<div style="background-color:#5DADE2"><b>Apartado 4</b></div>

```r
muestra = rgeom(1000,0.05) #Hacemos la muestra de 1000 intentos con una probabilidad de 0.05 para cada intento
frecuencia = prop.table(table(muestra)) # calculamos la frecuencia relativa de cada intento
probsTeo=data.frame(0:40,Prob=dgeom(1000,0.05)) # calculamos las probabilidades teoricas que debido a como esta programada la funcion dgeom los valores estan mal aproximados
merge(frecuencia, probsTeo, all = TRUE) # comparamos la frecuencia con las probabildades teoricas

#Interpretacion de resultados: como vemos la frecuencia relativa de la muestra se aproxima a los resultados que teoricamente deberiamos obtener pero al calcular las probabildades teoricas
# al estar mal programada la funcion dgeom no lo podemos ver reflejado en la tabla
```
