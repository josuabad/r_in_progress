```r
es_par = function (x){
    if (x %% 2 == 0){  # si el resto de x / 2 es =  0 entonces es par y se imprime "es par"
        "es par"
    }
    else{ # en otro caso (cuando el resto es -1) se imprime es impar
        "no es par"
    }
}
```

<div style="background-color:#5DADE2"><b>Apartado 5</b></div>

```r
sumar_hastaneg = function(vec){
    suma = 0 # Inicializa la suma en 0
    for(i in vec){ # Recorremos el vector
        if(i >= 0){ # Si el numero no es negativo se realiza la suma
            suma = suma + i
        }
        else { # Cuando el número es negativo se devuelve la suma total sin contar el propio número
            return (suma)
        }
    }
    return (suma) # Si no se encuentran negativos, se devuelve la suma cuando termine el for()
}
```

<div style="background-color:#5DADE2"><b>Apartado 4</b></div>

```r
frecuente = function(vec){
    for (i in vec){ # Recorremos el vector
        contador = 1 # Contador de números para repeticiones
        for (j in vec){ # Por cada elemento se recorre el vector otra vez
            if (i == j){ # Si hay elementos iguales, incrementa el contador
                contador = contador + 1
            }
        }
        if (contador >= 3){ # Si hay más de 3 elementos devuelve el mensaje
            return ("Hay un numero que se repite frecuentemente")
        }
    }
}
```

```r
frecuente(c(120,24,35,24,88,120,155,88,120,155))
```

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

```r
version_sign = function (vec){
    vector = c()
    for(i in vec){
        if(i > 0){
            vector = c(vector, 1)
        }
        else if (i == 0){
            vector = c(vector, 0)
        }
        else{
            vector = c(vector, -1)
        }
    }
    return (vector)
}
```

```r
version_sign(c(-4,2,0,3))
```

<div style="background-color:#5DADE2"><b>Apartado 4</b></div>

```r
reord_filas = function(mat){
    aux = c()
    for(i in 1:(nrow(mat)/2)){
        aux = mat[i,]
        mat[i,] = mat[nrow(mat)-i+1,]
        mat[nrow(mat)-i+1,] = aux    
    }
    return(mat)
}
```

```r
mat = matrix(1:12, nrow = 3, ncol = 4)
reord_filas(mat)
```

```r
# Vamos a escribir una función para que lo haga
elementosEnVector <- function(vector, elementos=0){
    # elementos(numeric) -> par(2) / impar(1) / todo(!(1, 2)) -> default(todo)
    if(elementos == 2) {
        elementosPares <- seq(from = 2, to = length(vector), by = 2)
        for(par in elementosPares) {
            print(vector[par])
        }
    } else if(elementos == 1) {
        elementosImpares <- seq(from = 1, to = length(vector), by = 2)
        for(impar in elementosImpares) {
            print(vector[impar])
        }
    } else {
        print(vector)
    }
}
```

La función desarrollada (elementosEnVector) muestra por pantalla los elementos de un vector. Admite 2 argumentos distintos: un vector y un numero. Dependiendo de lo que quiera mostrar;
1. Puede mostrar todos los elementos pares del vector con el número 2

```r
# Vamos a escribir una función para que lo haga
fibonacciSerie <- function(num) { # Introduce el número hasta el que enseñar
    fibonacci <- c(0, 1) # Necesitamos los valores base de la sucesión
    while(length(fibonacci) < num) {
        # Necesitamos un número extra que sea el que añadamos a los valores base (0, 1, ...)
        xtra <- fibonacci[length(fibonacci)] + fibonacci[length(fibonacci) - 1]
        fibonacci <- c(fibonacci, xtra)
    }
    return(fibonacci)
}
```

```r
fibonacciSerie(1000)
```
