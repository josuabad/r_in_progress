```r
sumar_hastaneg = function(vec){
    suma = 0 # Inicializa la suma en 0
    for(i in vec){ # Recorremos el vector
        if(i >= 0){ # Si el numero no es negativo se realiza la suma
            suma = suma + i
        }
        else { # Cuando el nmero es negativo se devuelve la suma total sin contar el propio nmero
            return (suma)
        }
    }
    return (suma) # Si no se encuentran negativos, se devuelve la suma cuando termine el for()
}
```

<div style="background-color:#5DADE2"><b>Apartado 4</b></div>

```r
frecuente = function(vec){
    for (i in vec){ # Recorremos el vector
        contador = 1 # Contador de nmeros para repeticiones
        for (j in vec){ # Por cada elemento se recorre el vector otra vez
            if (i == j){ # Si hay elementos iguales, incrementa el contador
                contador = contador + 1
            }
        }
        if (contador >= 3){ # Si hay ms de 3 elementos devuelve el mensaje
            return ("Hay un numero que se repite frecuentemente")
        }
    }
}
```

```r
frecuente(c(120,24,35,24,88,120,155,88,120,155))
```

<div style="background-color:#5DADE2"><b>Apartado 2</b></div>

```r
version_sign = function (vec){
    vector = c()
    for(i in vec){
        if(i > 0){
            vector = c(vector, 1)
        }
        else if (i == 0){
            vector = c(vector, 0)
        }
        else{
            vector = c(vector, -1)
        }
    }
    return (vector)
}
```

```r
version_sign(c(-4,2,0,3))
```

<div style="background-color:#5DADE2"><b>Apartado 4</b></div>

```r
reord_filas = function(mat){
    aux = c()
    for(i in 1:(nrow(mat)/2)){
        aux = mat[i,]
        mat[i,] = mat[nrow(mat)-i+1,]
        mat[nrow(mat)-i+1,] = aux    
    }
    return(mat)
}
```

```r
mat = matrix(1:12, nrow = 3, ncol = 4)
reord_filas(mat)
```

```r
# Vamos a escribir una funcin para que lo haga
fibonacciSerie <- function(num) { # Introduce el nmero hasta el que ensear
    fibonacci <- c(0, 1) # Necesitamos los valores base de la sucesin
    while(length(fibonacci) < num) {
        # Necesitamos un nmero extra que sea el que aadamos a los valores base (0, 1, ...)
        xtra <- fibonacci[length(fibonacci)] + fibonacci[length(fibonacci) - 1]
        fibonacci <- c(fibonacci, xtra)
    }
    return(fibonacci)
}
```

```r
fibonacciSerie(1000)
```
