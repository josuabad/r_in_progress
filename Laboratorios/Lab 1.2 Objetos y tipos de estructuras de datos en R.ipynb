## 1.1 Definir un vector

Un **vector** es un conjunto de valores bsicos del mismo tipo. La forma ms sencilla de crear vectores es a travs de la funcin `c()` que se usa para combinar (concatenar) valores.

```r
# Creamos un vector x numrico
x <- c(3, 5, 7)
x
```

```r
# Creamos un vector y numrico
y <- c(8, 9)
y
```

```r
# Creamos un vector a partir de dos vectores
c(x, y)
```

```r
# Creamos un vector z categrico o tipo cadena
z <- c("Hola", "Adios")
z
```

## 1.2 Generacin de secuencias numricas

Existen varias funciones que pemiten obtener **secuencias de nmeros**:

```r
# Creamos un vector x numrico con el operador :
x <- 1:5
x
```

## 1.6 Valores perdidos

Los **valores perdidos** aparecen normalmente cuando algn dato NO ha sido registrado. Este tipo de valores se registran como `NA` (abreviatura de Not Available). 

Supongamos que **tenemos registrado las alturas de 5 personas pero desconocemos la altura de la cuarta persona**. El vector sera registrado de la siguiente forma:

```r
# Creamos el vector numrico
altura <- c(165, 178, 184, NA, 175)
# Visualizamos
altura
```

**Es importante notar que cualquier operacin aritmtica sobre un vector que contiene algn `NA` dar como resultado otro `NA`**.

```r
# Si quisiramos calcular la altura media
mean(altura)
```

```r
# Para forzar a R a que ignore los valores perdidos se utliza la opcin na.rm = TRUE
mean(altura, na.rm = TRUE)
```

## 1.8 Factores

Los **factores** se utilizan para representar **datos categricos**. Se puede pensar en ellos como **vectores de enteros en los que cada entero tiene asociada una etiqueta (label)**. Los factores son muy importantes en la modelizacin de datos ya que los trata de forma especial. Utilizar factores con etiquetas es preferible a utilizar enteros porque las etiquetas son auto-descriptivas.

**Ejemplo** Supongamos que el vector sexo indica el `gnero` de un persona codificado como 0 si es hombre y 1 si es mujer.

```r
# Creamos el vector numrico
sexo <- c(0, 1, 1, 1, 0, 0, 1, 0, 1)
# Visualizamos
sexo
```

```r
# Obtenemos la frecuancia de cada valor
table(sexo)
```

El problema de introducir as los datos es que **NO queda reflejado la etiqueta** y por ende, el significado de los mismos. Para ello guardaremos los datos en una estructura **tipo factor**:

```r
# Transformamos la variable anterior a una variable tipo factor
sexo2 <- factor(sexo, labels = c("hombre", "mujer")); sexo2
```

```r
levels(sexo2)  # devuelve los niveles de un factor
```

```r
unclass(sexo2)  # representacin subyacente del factor
```

```r
# Obtenemos la frecuancia de cada valor
table(sexo2)
```

Veamos **otro ejemplo**, en el que inicialmente tenemos datos categricos. Los niveles se toman automticamente por orden alfabtico.

```r
# Definimos los valores de la variable
respuestas <- factor(c('si', 'si', 'no', 'si', 'si', 'no', 'no'), levels = c('si', 'no'))
# Visualizamos
table(respuestas)
respuestas
```

# 2. Matrices: objetos, estructuras y operaciones <a id="topic2"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>

## 2.1 Definir una matriz

Las **matrices** son la extensin natural de los vectores a **dos dimensiones**. Su **generalizacin a ms dimensiones** se llama **array**. Las matrices se pueden crear concatenando vectores con las funciones `cbind` o `rbind`:

```r
# Creamos un vector numrico x
x <- c(3, 7, 1, 8, 4)
x
```

```r
# Creamos un vector numrico y
y <- c(7, 5, 2, 1, 0)
y
```

```r
# Creamos una matriz concatenando los vectores por columnas
cbind(x, y)  
```

```r
# Creamos una matriz concatenando los vectores por filas
rbind(x, y)
```

Una matriz se puede crear con la funcin `matrix` donde el parmetro `nrow` indica el nmero de filas y `ncol` el nmero de columnas. **Por defecto, los valores se colocan por columnas**.

```r
# Creamos una matriz aplicando la funcin matrix
matrix(1:8, nrow = 2, ncol = 4)  # equivalente a matrix(1:8, nrow=2)
```

Los **nombres de los parmetros se pueden acortar** siempre y cuando no haya ambigedad, por lo que es habitual escribir.

```r
# Creamos una matriz aplicando la funcin matrix (nombres acortados)
matrix(1:8, nr = 2, nc = 4)
```

```r
# Creamos una matriz aplicando la funcin matrix (Si queremos indicar que los valores se escriban por filas)
matrix(1:8, nr = 2, byrow = TRUE)
```

## 2.2 Nombrar las filas y columnas en las matrices

Se pueden dar nombres a las filas y columnas de una matriz del siguiente modo:

```r
# Creamos una matriz aplicando la funcin matrix
x <- matrix(c(1, 2, 3, 11, 12, 13), nrow = 2, byrow = TRUE)
# Visualizamos
x
```

```r
# Creamos nombres de las filas con la funcin rownames
rownames(x) <- c("fila 1", "fila 2")
# Visualizamos
x 
```

```r
# Creamos nombres de las columnas con la funcin colnames
colnames(x) <- c("col 1", "col 2", "col 3")
# Visualizamos
x 
```

```r
# Obtenemos el mismo resultado si escribimos
colnames(x) <- paste("col", 1:ncol(x), sep=" ")
# Visualizamos
x
```

```r
# Internamente, las matrices son vectores con un atributo especial: la dimensin.
attributes(x)
```

## 2.3 Acceder a los elementos de una matriz

El **acceso a los elementos de una matriz** se realiza de forma anloga al acceso ya comentado para los vectores tal que,

```r
# Creamos una matriz
x <- matrix(1:6, 2, 3); x
```

```r
# Accedemos al valor de la 1 fila y columna
x[1, 1]
```

```r
x[2, ]  # segunda fila
```

```r
x[ ,2]  # segunda columna
```

```r
x[1, 1:2]  # primera fila, columnas 1 y 2
```

## 2.4 Ordenacin por filas y columnas

En ocasiones, interesar **ordenar los elementos de una matriz** por los valores de una determinada columna o fila. Por ejemplo, supongamos la matriz:

```r
# Creamos el vector numricos de valores
x <- c(79, 100, 116, 121, 52, 134, 123, 109, 80, 107, 66, 118); x
```

```r
# Transformamos el vector numrico en una matriz
x <- matrix(x, ncol=4, byrow=T); x
```

```r
# La matriz ordenada por los valores de la primera columna viene dada por
ii <- order(x[ ,1])
ii
```

```r
# Ordenacin columna 1
x[ii, ]  
```

```r
# De igual modo, si queremos ordenar por los valores de la cuarta columna:
ii <- order(x[ ,4]); x[ii, ]  # ordenacin columna 4
```

## 2.5 Operaciones fundamentales con matrices 

A continuacin se muestran algunas funciones que se pueden emplear con matrices para **realizar operaciones matriciales**:

## 2.6 Inversa de una matriz

Veamos un ejemplo de **cmo obtener la inversa de una matriz** utilizando algunas de las funciones antes mencionadas.

```r
# Creamos una matriz A
A <- matrix(c(2, 4, 0, 2), nrow = 2); A
```

```r
# Obtenemos el determinante de la matriz A
det(A)
```

```r
# Obtenemos la traspuesta de la matriz A
t(A)
```

```r
# Obtenemos la adjunta de la matriz A
adj <- matrix(c(2, -4, 0, 2), nrow = 2); adj
```

```r
# Obtenemos la inversa de la matriz A
inv <- (1/det(A))*adj
inv
```

```r
# Obtenemos la inversa de la matriz A DIRECTAMENTE
B <- solve(A)
B  
```

```r
A %*% B  # comprobamos que est bien calculada 
```

## 3.1 Definir una dataframe

Los **data.fames (marcos de datos) son el objeto ms habitual para el almacenamiento de datos**. En este tipo de objetos cada individuo de la muestra se corresponde con una fila y cada una de las variables con una columna. Para la creacin de estas estructuras se utiliza la funcin `data.frame()`.

Este tipo de estructuras son en apariencia **muy similares a las matrices**, con la ventaja de que permiten que los valores de las distintas columnas sean de tipos diferentes. Por ejemplo, supongamos que tenemos registrados los siguientes valores.

```r
# Creamos una serie de vectores numricos y categricos
Producto <- c("Zumo", "Queso", "Yogourt")
Seccion <- c("Bebidas", "Lcteos", "Lcteos")
Unidades <- c(2, 1, 10)
```

```r
# Los valores anteriores se podran guardar en una nica matriz
x <- cbind(Producto, Seccion, Unidades)
class(x)
```

```r
# Visualizamos los resultados
x
```

Sin embargo, el resultado anterior **NO es satisfactorio** ya que todos los valores se han **transformado en caracteres**. Una solucin mejor es utilizar un `data.frame`, con lo cual se **mantiene el tipo original de las variables**.

```r
# Creamos el data frame
lista.compra <- data.frame(Producto, Seccion, Unidades)
# Vemos la clase
class(lista.compra)
```

```r
# Visualizamos
lista.compra
```

## 3.4 Manipulacin de dataframes

A menudo se dice que gran parte del anlisis de datos se gasta en el **proceso de limpieza y preparacin de los datos**. ### 3.4.1 La librera dplyr

El paquete `dplyr` fue desarrollado por Hadley Wickham de RStudio y es un **versin optimizada de su paquete plyr**. La librera dplyr **NO proporciona ninguna nueva funcionalidad a R per se**, en el sentido que todo aquello que podemos hacer con dplyr lo podramos hacer con la sintaxis bsica de R. En particular, contiene un conjunto de herramientas (funciones) sumamente intuitivas y poderosas para una **rpida y fcil manipulacin y procesamiento de los datos**. 

Una **importante contribucin del paquete dplyr** es que **proporciona una "gramtica"** (particularmente verbos) para la manipulacin y operaciones con dataframes. Con esta gramtica podemos comunicar mediante nuestro cdigo que es lo que estamos haciendo en los dataframes a otras personas (asumiendo que conozcan la gramtica). 

Adems, cabe destacar que **las funciones del paquete dplyr son muy rpidas**, puesto que estn implementadas con el lenguaje C++. Todas las funciones de dicha librera tienen en comn una serie de argumentos. En particular,

+ El **primer argumento es el dataframe**.
+ Los **otros argumentos describen qu hacer con el dataframe** especificado en el primer argumento, podemos referirnos a las columnas en el data frame directamente sin utilizar el operador `$`, es decir slo con el nombre de la columna/variable.
+ El **valor de retorno es un nuevo dataframe**.
+ Los data frames **deben estar bien organizados/estructurados**, es decir debe existir una observacin por columna y, cada columna representar una variable, medida o caracterstica de esa observacin. 

Las posibilidades de dplyr son muy extensas, os recomiendo que miris el captulo 5 del libro [R for Data Science](https://r4ds.had.co.nz/transform.html). Veamos simplemente un pequeo ejemplo de su uso siguiendo el dataset anterior. Algunas de los **principales "verbos"** de la librera dplyr son:

```r
# Instalamos la librera dplyr
install.packages("dplyr")
```

```r
# Esta funcin tambin sirve para cargar las libreras es equivalente a 'library'
require(dplyr)
```

+ `select`: selecciona el subconjunto del dataset que cumpla ciertas caractersticas.

```r
# elige un subconjunto de las variables (columnas) 
select(iris, Sepal.Length, Sepal.Width)
```

+ `filter`: devuelve un conjunto de filas segn una o varias condiciones lgicas.

```r
# selecciona todos los lirios de la especie setosa
filter(iris, Species=='setosa')
```

+ `arrange`: reordena filas de un dataframe.

```r
# ordena de acuerdo con la longitud del spalo
arrange(iris, Sepal.Length)
```

+ `%>%` : el operador "pipe" es usado para conectar mltiples acciones en una nica "pipeline" (tubera). Por ejemplo, si queremos seleccionar las variables que contienen las medidas del ptalo, seleccionar los lirios para los que la longitud del ptalo es mayor que 4 mm y ordenarlos de menor a mayor longitud del ptalo.

```r
# Sintaxis en cadena
iris %>%
  select(contains('Petal'))  %>%
  filter(Petal.Length > 4)   %>%
  arrange(Petal.Length)
```

**Nota.** El trmino a la izquierda de cada operador %>% es el primer argumento del trmino de la derecha (y por lo tanto el primer argumento se puede omitir).

```r
# Podemos comparar el cdigo anterior con el cdigo anidado habitual:
arrange(filter(select(iris, contains('Petal')), Petal.Length > 4), Petal.Length)
```

+ `mutate`: aade nuevas variables/columnas o transforma variables existentes. Por ejemplo, creamos una variable que corresponde al cociente entre la anchura y la longitud del ptalo (que podra corresponder a algn aspecto de la forma del ptalo).

```r
iris %>%
  mutate(forma = Petal.Width/Petal.Length)
```

+ `summarise`: para aplicar comandos a variables. Normalmente se usa en combinacin con `group_by` de manera que se calculen estadsticos para subgrupos de observaciones. En el siguiente ejemplo se calcula la media de la longitud del ptalo para los lirios de cada una de las especies.

```r
iris %>%
  group_by(Species) %>%
  summarise(mean(Petal.Length))
```

# 4. Listas: objetos, estructuras y operaciones <a id="topic4"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>

## 4.1 Definir una lista

Las **listas** son **colecciones ordenadas de cualquier tipo de objetos** (en R las listas son un tipo especial de vectores). As, mientras que los **elementos de los vectores, matrices y arrays deben ser del mismo tipo**, en el caso de las **listas se pueden tener elementos de tipos distintos**. Luego, por esos motivos simplemente pondremos un pequeo ejemplo ya que la mayora de cuestiones vistas en cada uno de los objetos anteriores las podemos aplicar en el caso de las listas.

```r
# Creamos vectores numricos, categricos y una matriz
x <- c(1, 2, 3, 4); x
y <- c("Hombre", "Mujer"); y
z <- matrix(1:12, ncol = 4); z
```

```r
# Creamos una lista a partir de los objetos previos
datos <- list(A=x, B=y, C=z)
datos
```

Supongamos que queremos **guardar en una lista un vector, su nombre, su media, y su vector de sumas acumuladas**. En este caso, tendramos que hacerlo en forma de lista heterognea usando la funcin `list`.

```r
# Creamos un vector numrico
x=c(1,2,-3,-4,5,6)
```

```r
# Creamos una lista a partir de valores del vector
listas=list(nombre="x", vector=x, media=mean(x), sumas=cumsum(x))
# Visualizamos
listas
```

Observar la **sintaxis de la funcin $\tt list$**: le hemos metido como argumento los diferentes objetos que van a formar la lista heterognea, poniendo a cada uno un nombre adecuado. Este nombre es interno de la `lista`: por ejemplo, pese a que `listas` contiene un objeto llamado `sumas`, en el entorno de trabajo de R no tenemos definida ninguna variable con ese nombre (a no ser que la hayamos definido previamente durante la sesin).

```r
# El objeto sumas NO existe
sumas
```

## 4.2 Acceder a los elementos de una lista

Para referirnos o usar una componente concreta de una `lista`, tenemos que aadir al nombre de la `lista` el sufijo formado por un signo `$` y el nombre de la componente.

```r
# Accedemos a la 1 variable de la lista
listas$nombre
```

```r
# Accedemos a la 2 variable de la lista
listas$vector
```

```r
# Accedemos a la 3 variable de la lista
listas$media
```

Tambin podemos **indicar el objeto** por su posicin en la lista **usando dobles corchetes** `[[ ]]`. Si usamos slo un par de corchetes, como en los vectores, lo que obtenemos es una lista formada por esa nica componente, no el objeto que forma la componente.

```r
# Accedemos a la 1 variable de la lista
listas[[1]]
```

```r
# Accedemos a la 4 variable de la lista
listas[[4]]  # Esto es un vector
```

```r
3*listas[[4]] # Y podemos operar con l
```

```r
listas[4]   # Esto es una lista, no un vector
```

```r
3*listas[4]   # NO podemos operar con l
```

# 5. Contenido adicional: la familia de funciones $\tt apply$ <a id="topic5"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>

La familia de **funciones de tipo $\tt apply$** es usada para *aplicar una funcin a cada elemento de una estructura de datos dada*. En particular, es usada para aplicar funciones sobre los objetos que acabamos de ver como son las matrices, los data frames, los arrays o las listas. Con esta familia de funciones podemos automatizar tareas complejas usando pocas lneas de cdigo y es una de las caractersticas distintivas de R como lenguaje de programacin.

**La familia de funciones apply NO slo recibe datos como argumentos, tambin recibe funciones**. Para entender ms fcilmente el uso de la familia apply, recordemos la vectorizacin de operaciones. Hay operaciones que, si las aplicamos a un vector, son aplicadas a todos sus elementos.

```r
# Creamos un vector x con nmeros del 1 al 10
x=1:10
x
```

```r
# Esto sera un ejemplo de vectorizacin de operaciones. En este caso, de la operacin potencia.
x^2
```

Lo anterior es, generalmente, preferible a escribir una operacin para cada elemento. Luego, podemos decir que al **vectorizar** estamos aplicando una funcin a cada elemento de un vector. **La familia de funciones ${\tt apply}$ nos permite implementar esto en estructuras de datos distintas a los vectores**. 

<mark>**Todas las funciones de esta familia tienen una caracterstica en comn reciben como argumentos a un objeto y al menos una funcin**</mark>. 

Hasta ahora, todas las funciones que hemos usado han recibido como argumentos estructuras de datos, sean vectores, dataframes o matrices, por ejemplo. Las funciones de la familia ${\tt apply}$ tienen la particularidad que pueden recibir a otra funcin como un argumento.

La familia ${\tt apply}$ est formada por las siguientes 8 funciones (ordenadas alfabticamente):

+ ${\tt apply()}$
+ ${\tt eapply()}$
+ ${\tt lapply()}$
+ ${\tt mapply()}$
+ ${\tt rapply()}$
+ ${\tt sapply()}$
+ ${\tt tapply()}$
+ ${\tt vapply()}$

Nosotros repasaremos la funcin ms general y de uso comn de esta familia: la funcin ${\tt apply()}$. Esta funcin nos permitir solucionar casi todos los problemas que nos encontremos. Adems, conociendo su uso, las dems funciones de la familia ${\tt apply}$ sern relativamente fciles de entender. **La sintaxis de esta funcin es**:

```r
# apply(X, MARGIN, FUN, ...)
```

+ `X`: una matriz o cualquier elemento que podamos convertir a una matriz, es decir, un dataframe o un vector normalmente.
+ `MARGIN`: la dimensin que agrupar los elementos de la matriz `X`, para aplicarles una funcin. Son identificadas con nmeros, 1 son filas y 2 son colummnas.
+ `FUN`: la funcin que aplicaremos a la matriz `X` en su dimensin `MARGIN`.
+ `...`: argumentos opcionales que sern usados por `FUN`.

 Por ejemplo, podemos usar ${\tt apply()}$ para obtener la suma de los elementos de una matriz por filas.

```r
# Creamos una matriz de cuatro filas tal que
matriz <- matrix(1:16, nrow = 4)
# Visualizamos la matriz creada 
print(matriz)
```

```r
# Aplicamos la funcin apply para obtener la suma de los elementos de una matriz por filas
apply(X = matriz, MARGIN = 1, FUN = sum) # En el argumento FUN la funcin va sin parntesis ()
```

```r
# Probamos otra alternativa sin la funcin apply
c(sum(matriz[1, ]),sum(matriz[2, ]),sum(matriz[3, ]),sum(matriz[4, ])) # Equivalente
```

```r
# Aplicamos la funcin apply para obtener la suma de los elementos de una matriz por columnas
apply(X = matriz, MARGIN = 2, FUN = sum)
```

Recordar que **podemos dar como argumento cualquier nombre de funcin, siempre y cuando sta acepte vectores como argumentos** (los elementos de una matrix o dataframe son vectores). Probemos cambiando el argumento ${\tt FUN}$. 

```r
# Aplicamos la funcin apply para obtener la media de cada fila
apply(matriz, 1, mean) 
```

```r
# Aplicamos la funcin apply para obtener la media de cada columna
apply(matriz, 2, mean) 
```

```r
# Aplicamos la funcin apply para obtener el mximo de cada fila
apply(matriz, 1, FUN = max) # Se ejecuta sin necesidad de especificar argumentos
```

<mark>**Por lo tanto, ${\tt apply}$ slo puede ser usado con funciones que esperan vectores numricos como argumentos.**</mark> Si proporcionamos un argumento invlido, la funcin NO se ejecutar y ${\tt apply}$ fallar.
