# 1. Breve introducción al lenguaje de programación de R <a id="r"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


R es un lenguaje y entorno de programación, creado en 1993 por Ross Ihaka y Robert Gentleman del Departamento de Estadística de la Universidad de Auckland, cuya característica principal es que **forma un entorno de programación idóneo para el análisis estadístico y la ciencia de datos**. En su aspecto R puede considerarse como otra implementación del lenguaje de programación S, con la particularidad de que es un software GNU, General Public Licenc (conjunto de programas desarrollados por la Free Software Foundation), es decir, **de uso libre**.

Una **diferencia fundamental de la filosofía** de R, y también de la de S, con el resto del software matemático y estadístico es el **uso de *objetos*** (variables, variables indexadas, cadenas de caracteres, funciones, etc.) **como entidad básica**. Cualquier expresión evaluada por R se realiza en una serie de pasos, con unos resultados intermedios que se van almacenando en objetos, para ser observados o analizados posteriormente, de tal manera que se puede hacer un análisis sin necesidad de mostrar su resultado inmediatamente produciendo unas salidas mínimas.

Por tanto, cada **objeto pertenece a una clase, de forma que las funciones pueden tener comportamientos diferentes en función de la clase a la que pertenece su objeto argumento**. Por ejemplo, no se comporta igual una función cuando su argumento es un vector que cuando es un fichero de datos u otra función. 

*R* **está disponible en varios formatos**: en **código fuente está escrito principalmente en *C*** (y algunas rutinas en *Fortran*), esencialmente para máquinas *Unix* y *Linux*, o como archivos binarios precompilados para *Windows*, *Linux* (Debian, Mandrake, RedHat, SuSe), *Macintosh* y *Alpha Unix*.

Junto con *R* se incluyen ocho bibliotecas o paquetes (llamadas bibliotecas estándar) pero otros muchos están disponibles a través del repositorio público [Comprehensive R Archive Network (CRAN)](https://cran.r-project.org/). 

Actualmente se encuentran **disponibles más de 19000 librerias (packages)** desarrollados en *R*, que cubren multitud de campos como, por ejemplo: <mark>Bayesian Inference, Probability Distributions, Functional Data Analysis, Machine Learning & Statistical Learning, etc.<mark> 
    
Esto es lo que define *R* como un **entorno vivo**, que se actualiza con frecuencia y que está abierto a la mejora continua. Puede consultarse una **clasificación de las diferentes librerías** por **topics** en [Task Views](https://cran.r-project.org/). Os recomiendo que ojeéis ese enlace para familiarizaros con la multitud de librerias disponibles en *R*.
    
Os dejo por aquí algunos enlaces a conferencias, congresos y/o seminarios vinculados a R: [International R User Conference](https://www.r-project.org/conferences/), [R-Hispano](http://r-es.org/), [R-Madrid](http://madrid.r-es.org/).

# 3. Realizando el set-up de R en la plataforma Anaconda <a id="setup"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Como hemos comentado anteriormente, básicamente **R** es un lenguaje que **permite manipular objetos estadísticos** y crear gráficos de alta calidad. Es a la vez un **entorno interactivo** y un **lenguaje de programación interpretado con funciones orientadas a objetos**. En esta sección veremos cómo instalar R en el entorno de trabajo de la plataforma de Data Science **Anaconda**. Tendremos en cuenta dos procedimientos dependiendo del sistema operativo que se utilice. La versión actual al escribir el presente documento es **R-4.3.1** publicado el **16-06-2023**.

## 3.1 Descargar e instalar Anaconda

En primer lugar, tenemos que descargar e instalar Anaconda desde su propia web considerando el sistema operativo propio. Ver [https://www.anaconda.com/](https://www.anaconda.com/).


## 3.2 Cargar R en el entorno Anaconda (Windows)

Si has instalado Anaconda, al ejecutar este programa se abrirá una ventana como la mostrada en la siguiente imagen:

Esta página nos permite navegar por el disco duro hasta la carpeta en la que tengamos nuestros *Jupyter notebooks*. Estos *cuadernos* o *notebooks* contendrán el código R, las visualizaciones que se hayan generado, etc. Tened en cuenta que por defecto, en **Anaconda** viene cargado el lenguaje de programación *Python*. Luego, si abrimos un *Jupiter Notebook*, las celdas de código leeran *Python*. Si queremos integrar *R* en los *Jupyter Notebooks*, tendremos que relizar un sencillo paso.

Algunas de las librerías que contienen **funciones y algoritmos que vamos a tener que utilizar a lo largo del curso necesitan que, al menos, esté instalada la versión R 4.+** y en Navigator Anaconda la versión de R es la 3.6.1. Eso podrá generaros problemas a la hora de tratar con funciones de determinadas librerías. Luego, lo único que tenéis que hacer es un upgrade de la versión de R. ## 3.3 Cargar R en el entorno Anaconda (MacOS) 

En el caso de aquellos que tengáis el sistema operativo MacOS, una vez instalado Anaconda, al abrirlo se mostrará la siguiente imagen. 

Como podéis ver es exactamente igual que en el caso de Windows. La **única diferencia** es en el procedimiento que utilizamos para crear el **entorno de R** y además, que en este caso, NO necesitamos luego actualizar la versión de R ya que lo llevamos a cabo directamente a la hora de crear el nuevo entorno. Veamos que tenéis que hacer. Ojo! Tenéis que cerrar el programa Anaconda.

#### <mark> Crear nuevo environment en Anaconda para poder utilizar R (MacOS)<mark>

Utilizad los **siguientes comandos en el propio terminal de vuestro pc** (no en la consola de R ni en Jupyter Notebook).
        
1. Una versión actualizada de R (>=4.0) está disponible a través del canal de conda-forge, así que primero tenéis que añadir el canal `conda config --add channels conda-forge`.
    
2. Vamos a instalar R a través de conda-forge y no a través del canal por defecto, por lo que debemos establecer su prioridad sobre el canal por defecto `conda config --set channel_priority strict`.

3. Comprobamos si se añade una versión actualizada de R en el espacio de búsqueda de conda `conda search r-base`.   
    
4. Luego, tenemos que crear un nuevo entorno conda, que ayudará a depurar los problemas específicos del paquete y de compatibilidad sin interrumpir el entorno base `conda create -n R-Server python=3.6`.
    
5. Vamos a activar el entorno conda recién creado `conda activate R-Server`.
    
6. Instalamos R 4.+ en Anaconda `conda install -c conda-forge r-base`.
    
7. El nuevo entorno aún no permite utilizar los Jupyter Notebooks con el kernel de R 4.+. Para ello, necesitáis instalar la librería $\tt IRkernel$ y Jupyter notebook. Para ello, abrir Anaconda, os vais a la pestaña de environments a la izquierda, cargáis el nuevo entorno, volvéis a Home, y tenéis que darle a Install en la aplicación de Jupyter Notebooks. Se podrá demorar unos minutos su instalación.

8. Luego, le dais a Install en la aplicación de RStudio en el nuevo entorno. Se podrá demorar unos minutos su instalación.
    
9. Una vez instalado RStudio, tenéis que abrirlo e instalar la librería $\tt IRkernel$ para proporcionar a Anaconda el kernel de R 4.+. 
    
10. Finalmente, tenéis que introducir en la consola de RStudio el siguiente comando,`IRkernel::installspec()`, para habilitar que los Jupyter Notebooks puedan leer código en el nuevo kernel de R 4.+. 
    
11. Listo. Solamente os tenéis que ir ahora a la pestañana de environments y darle al botón del `Play` en el entorno creado de R y seleccionar la opción de **Open with Jupyter Notebook** tal como se muestra en la siguiente imagen. 
    


Tras realizar esa operación, ya podremos crear un Jupiter Notebook que nos permita incluir en las celdas de código la sintaxis del lenguaje de *R*. Para ello, tenemos que hacer clic en la pestaña *New* que se ve en la esquina superior derecha de la página principal y, a continuación, otro clic en la opción **R** del menú desplegable que se habrá mostrado. 

Esto abrirá en nuestro navegador otra ventana conteniendo nuestro **cuaderno nuevo** inicialmente vacío. Un clic en el texto *Untitled* (en la siguiente imagen) nos permitirá dar un nombre al cuaderno (será el nombre del fichero en el que se guarde).

# 4. Primeros pasos con el lenguaje de programación de R <a id="pasos"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


¡Comencemos! El **signo *#*** indica la introducción de un *comentario*. Por ejemplo, la siguiente instrucción sumará 3 y 4 e ignorará el comentario:

```r
# Aquí podemos incluir comentarios
3+4
```

El **nombre de un objeto de *R***, sea un vector, una lista, una función, etc., puede ser cualquier cadena alfanumérica formada por letras (teniendo en cuenta que *R* distingue entre mayúsculas y minúsculas), dígitos del 0 al 9 (no como primer carácter) y el signo . (punto), sin limitación de número de caracteres.

```r
# Por ejemplo, Exp1289, muestra.ini o muestra.ini.ajuste son nombres válidos
```

**R** tiene **palabras reservadas** como son los nombres de las instrucciones utilizadas en el lenguaje de programación (*break*, *for*, *function*, *if*, *in*, *next*, *repeat*, *return*, *while*) y los de las funciones incorporadas en el propio entorno del programa, que **no se pueden usar como identificador de objetos**. En el caso de intentar redefinir una función ya utilizada por el programa, **R** advierte de la duplicidad de definiciones.

```r
x=2
print(x)
```

```r
for=2
```

Como hemos visto, **R** es un lenguaje interpretado con funciones orientadas a objetos. Las **instrucciones básicas son expresiones o asignaciones**. Para **realizar una asignación se pueden utilizar los signos <-, -> o *=***.

```r
n<-5*2+sqrt(144)
m=4^-0.5
n+m->p
```

Para **visualizar el contenido de un objeto** sólo es necesario **escribir su nombre**.

```r
n
```

```r
m; p
```

```r
(x <- log(7)) # (...) hace que se muestre el valor de x
```

```r
summary # si incluímos en nombre de una función predeterminada en R la salida es la siguiente
```

Los **comandos *objects()*** y ***ls()*** visualizan el listado de objetos presentes en el actual espacio o área de trabajo (workspace).

```r
objects()
```

```r
ls()
```

La mayoría de las **librerías disponibles** en R han de ser **cargados antes de que se los pueda utilizar**. Por ejemplo, para cargar el paquete *caret*, que contiene diferentes algoritmos de machine learning, se puede ejecutar la siguiente instrucción. ¡Ojo¡ Previamente, si no la teneís instalada, tendrías que hacerlo con el comando *install.packages("nombre de la librería")*. <mark>**Importante:** se instala solo 1 vez la librería y se cargan cada vez que abramos el notebook y/o la sesión de trabajo.<mark>
    
Es decir, es importante que tengas en mente que debes hacer una llamada a **library()** cada que inicies una sesión en R. Aunque hayas importado las funciones de un paquete con anterioridad, las sesiones de R se inician “limpias”, sólo con los objetos y funciones de base.

```r
install.packages("caret")
```

```r
library(caret)
```

A partir de ese momento, podríamos utilizar las funciones que contiene la librería *caret*. 

También podemos **ver qué paquetes tenemos ya instalados** ejecutando la función **installed.packages()** sin ningún argumento. Una instalación nueva de R tiene pocos paquetes instalados, pero esta lista puede crecer considerablemente con el tiempo. Recordad que podéis **consultar una clasificación de las diferentes librerías** por **topics** en [Task Views](https://cran.r-project.org/). Os recomiendo que ojeéis ese enlace para familiarizaros con la multitud de librerias disponibles en *R*.

```r
# installed.packages()
```

Sólo tienes que escribir la función en la consola y ejecutarla

La función **list.dirs()**, también sin argumentos te dará una lista de los directorios dentro del directorio de trabajo.

```r
# Ver archivos
list.files()
```

# 5. Documentación y búsqueda de ayuda en R <a id="ayuda"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


En el propio repositorio de **CRAN** disponéis de manuales en formato *html*, *pdf* y *epub* a los cuales se puede acceder vía online a través del siguiente enlace [Comprehensive R Archive Network (CRAN)](https://cran.r-project.org/). En el menú de la izquierda. En la sección de *Documentation* tenéis que hacer clic sobre *Manuals* y veréis lo siguiente.

Además, **R** dispone de una ayuda muy completa sobre todas las funciones, procedimientos y elementos que configuran el lenguaje. Se puede acceder a información específica sobre las mismas con el comando ***help*** o mediante **‘?’**. Se os abrirá un desplegable que resulta muy útil ya que nos proporciona una información muy completa sobre la **función**: <mark>librería a la que pertenece, parámetros que considera la función, ejemplos de uso o referencias bibliográficas en las que se ha basado para su implementación<mark>.


```r
# Ejemplo ayuda de la función summary
?summary
```

Para visualizar un **listado de todas las funciones que conienten ese comando** utilizaremos la función **apropos**:

```r
apropos("summary")
```

Para **reproducir un ejemplo** empleando la función en la que estamos interesados para ver el **output de la función** 

```r
# Salidas de la función summary
example(summary)
```

Para **obtener información NO sobre las funciones sino las librerías** hay que utilizar las funciones ***library*** y ***help*** conjuntamente. Nuevamente, observamos una información muy completa de la librería, por ejemplo, *caret*: <mark>versión, autores, dependencias o el propio índice de todas las funciones que contiene la misma.<mark>

```r
library(help="caret")
```

También es posible **obtener ayuda sobre diferentes temas** mediante la función ***help.search***. **R** buscará ayuda sobre el tema escogido en todos las librerías instaladas. **Ejemplo:** si nosotros queremos saber **qué librerías** y **funciones**  contienen **algoritmos** relacionados con los algoritmos de **machine learning de redes neuronales** tendríamos que indicarle lo siguiente. Importante escribirlo en inglés. 

```r
help.search("neural networks")
```

Existen muchas **listas de ayuda de correo para **R****. Se recomiendan, por un lado, las distintas listas que se presentan en la página web de R (hacer clic sobre *Mailing lists*) en [http://www.r-project.org](http://www.r-project.org) y por otro lado la **lista de ayuda en castellano** (ver la página web [https://stat.ethz.ch/mailman/listinfo/r-help-es)](https://stat.ethz.ch/mailman/listinfo/r-help-es)).

La función ***RSiteSearch*** permite buscar palabras de interés entre todos los mensajes enviados a las amplias listas de ayuda de correo electrónico del propio repositorio **CRAN** de R. **Ejemplo:** si estamos buscando información sobre la prueba de Kolmogorov-Smirnov en R, podemos ejecutar:

```r
RSiteSearch("Kolmogorov Smirnov test")
```

También podemos encontrar ayuda en **foros como Stack Overflow** [https://stackoverflow.com/](https://stackoverflow.com/). El éxito en estos foros se mide en términos de que alguien responda a tu pregunta, así que es imprescindible aprender a formular preguntas de modo eficiente. De hecho, tienen una guía sobre cómo se hace una buena pregunta.

Finalmente, tendríamos la web de **R Documentation** en la que aparece recogida la documentación sobre todas y cada una de las funciones publicadas en R: [R Documentation](https://www.rdocumentation.org/).
