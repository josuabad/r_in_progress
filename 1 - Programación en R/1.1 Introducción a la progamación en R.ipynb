# 1. Breve introduccin al lenguaje de programacin de R <a id="r"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


R es un lenguaje y entorno de programacin, creado en 1993 por Ross Ihaka y Robert Gentleman del Departamento de Estadstica de la Universidad de Auckland, cuya caracterstica principal es que **forma un entorno de programacin idneo para el anlisis estadstico y la ciencia de datos**. En su aspecto R puede considerarse como otra implementacin del lenguaje de programacin S, con la particularidad de que es un software GNU, General Public Licenc (conjunto de programas desarrollados por la Free Software Foundation), es decir, **de uso libre**.

Una **diferencia fundamental de la filosofa** de R, y tambin de la de S, con el resto del software matemtico y estadstico es el **uso de *objetos*** (variables, variables indexadas, cadenas de caracteres, funciones, etc.) **como entidad bsica**. Cualquier expresin evaluada por R se realiza en una serie de pasos, con unos resultados intermedios que se van almacenando en objetos, para ser observados o analizados posteriormente, de tal manera que se puede hacer un anlisis sin necesidad de mostrar su resultado inmediatamente produciendo unas salidas mnimas.

Por tanto, cada **objeto pertenece a una clase, de forma que las funciones pueden tener comportamientos diferentes en funcin de la clase a la que pertenece su objeto argumento**. Por ejemplo, no se comporta igual una funcin cuando su argumento es un vector que cuando es un fichero de datos u otra funcin. 

*R* **est disponible en varios formatos**: en **cdigo fuente est escrito principalmente en *C*** (y algunas rutinas en *Fortran*), esencialmente para mquinas *Unix* y *Linux*, o como archivos binarios precompilados para *Windows*, *Linux* (Debian, Mandrake, RedHat, SuSe), *Macintosh* y *Alpha Unix*.

Junto con *R* se incluyen ocho bibliotecas o paquetes (llamadas bibliotecas estndar) pero otros muchos estn disponibles a travs del repositorio pblico [Comprehensive R Archive Network (CRAN)](https://cran.r-project.org/). 

Actualmente se encuentran **disponibles ms de 19000 librerias (packages)** desarrollados en *R*, que cubren multitud de campos como, por ejemplo: <mark>Bayesian Inference, Probability Distributions, Functional Data Analysis, Machine Learning & Statistical Learning, etc.<mark> 
    
Esto es lo que define *R* como un **entorno vivo**, que se actualiza con frecuencia y que est abierto a la mejora continua. Puede consultarse una **clasificacin de las diferentes libreras** por **topics** en [Task Views](https://cran.r-project.org/). Os recomiendo que ojeis ese enlace para familiarizaros con la multitud de librerias disponibles en *R*.
    
Os dejo por aqu algunos enlaces a conferencias, congresos y/o seminarios vinculados a R: [International R User Conference](https://www.r-project.org/conferences/), [R-Hispano](http://r-es.org/), [R-Madrid](http://madrid.r-es.org/).

# 5. Documentacin y bsqueda de ayuda en R <a id="ayuda"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


En el propio repositorio de **CRAN** disponis de manuales en formato *html*, *pdf* y *epub* a los cuales se puede acceder va online a travs del siguiente enlace [Comprehensive R Archive Network (CRAN)](https://cran.r-project.org/). En el men de la izquierda. En la seccin de *Documentation* tenis que hacer clic sobre *Manuals* y veris lo siguiente.

Adems, **R** dispone de una ayuda muy completa sobre todas las funciones, procedimientos y elementos que configuran el lenguaje. Se puede acceder a informacin especfica sobre las mismas con el comando ***help*** o mediante **?**. Se os abrir un desplegable que resulta muy til ya que nos proporciona una informacin muy completa sobre la **funcin**: <mark>librera a la que pertenece, parmetros que considera la funcin, ejemplos de uso o referencias bibliogrficas en las que se ha basado para su implementacin<mark>.


```r
# Ejemplo ayuda de la funcin summary
?summary
```

Para visualizar un **listado de todas las funciones que conienten ese comando** utilizaremos la funcin **apropos**:

```r
apropos("summary")
```

Para **reproducir un ejemplo** empleando la funcin en la que estamos interesados para ver el **output de la funcin** 

```r
# Salidas de la funcin summary
example(summary)
```

Para **obtener informacin NO sobre las funciones sino las libreras** hay que utilizar las funciones ***library*** y ***help*** conjuntamente. Nuevamente, observamos una informacin muy completa de la librera, por ejemplo, *caret*: <mark>versin, autores, dependencias o el propio ndice de todas las funciones que contiene la misma.<mark>

```r
library(help="caret")
```

Tambin es posible **obtener ayuda sobre diferentes temas** mediante la funcin ***help.search***. **R** buscar ayuda sobre el tema escogido en todos las libreras instaladas. **Ejemplo:** si nosotros queremos saber **qu libreras** y **funciones**  contienen **algoritmos** relacionados con los algoritmos de **machine learning de redes neuronales** tendramos que indicarle lo siguiente. Importante escribirlo en ingls. 

```r
help.search("neural networks")
```
