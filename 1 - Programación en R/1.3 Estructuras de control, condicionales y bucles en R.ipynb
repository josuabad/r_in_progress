# 1. Primeros pasos a la hora de programar una funcin en R <a id="topic1"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


El lenguaje R permite al usuario definir sus **propias funciones**. Una **funcin** permite escribir un bloque de cdigo y ejecutarlo para distintos datos. Una funcin puede considerarse un subprograma que resuelve una subtarea. Un motivo para utilizar funciones es que permiten estructurar u organizar el cdigo de un programa. Cuando hay que resolver un problema complejo, en lugar de intentar solucionarlo mediante un programa muy extenso es mejor **descomponerlo en subproblemas**.

En notebooks previos ya hemos usado algunas funciones de R, en este notebook, vamos a aprender a crear nuestras propias funciones. El esquema de una funcin es el que sigue:

```r
nombre_funcin = function(arg1, arg2, ... ){expresin}
```

+ En la expresin anterior `arg1, arg2, ...` son los argumentos de entrada (tambin llamados parmetros).

+ La `expresin` est compuesta de comandos que utilizan los argumentos de entrada para dar la **salida** deseada.

+ La **salida de una funcin** puese ser un `nmero, un vector, una grafica, un mensaje, etc`.

**Ejemplo 1 <mark>[funcin suma]</mark>** Vamos a empezar tratando de obtener una funcin que sume dos nmeros. Por ejemplo el 3 y el 4. Si no creamos una funcin a cada paso tendramos que ir utilizando el siguiente cdigo **manualmente**.

```r
# Sumar nmeros manualmente
3+4
```

Si queremos **automatizar** dicho proceso e **integrarlo en una funcin** tendramos que operar de la siguiente manera. `Consideramos 2 nmeros`, por tanto, `2 argumentos` a los que le vamos a llamar **'a'** y **'b'**. Teniendo en cuenta los principales elementos de una funcin descritos previamente tendramos lo siguiente. Vamos a llamar a nuestra funcin `suma`.

```r
# Sumar nmeros a travs de una funcin creada por nosotros
suma = function(a,b){
    out = a+b
}
```

```r
# Aplicamos la funcin para a=3 y b=4
suma(3,4)
```

Se ha ejecutado **pero NO hemos obtenido ningn tipo de objeto de salida**. Para ello, deberamos de haber incluido en nuestra funcin como siempre el propio objeto para que nos saque la informacin que contiene al ejecutarlo. 

```r
# Sumar nmeros a travs de una funcin creada por nosotros (con salida)
suma = function(a,b){
    out = a+b
    out
}
```

```r
# Aplicamos la funcin para a=3 y b=4 (con salida)
suma(3,4)
```

**<u>Descripcin funcin suma</u>.-** Para crear una funcin **se usa la palabra reservada** `function`. A continuacin se **enumeran los parmetros formales** de la funcin separados por comas y encerrados entre parntesis (en el ejemplo los parmetros formales son `a` y `b`). Despus viene el **cdigo de la funcin**. Puesto que el cdigo del ejemplo tiene ms de una instruccin, hay que encerrarlo entre llaves. Con esto se crea un **objeto de salida** `out` donde se almacena la operacin anterior. Finalmente, la funcin creada se asigna al objeto `suma`.

**Ejemplo 2 <mark>[funcin mximo]</mark>** Ahora vamos a crear una funcin que calcula el mximo de sus dos argumentos.

```r
# Obtener el valor mximo de 2 valores
maximo <- function(a, b) {
  if (a > b) {
    out <- a
  } else {
    out <- b
  }
  out
}
```

```r
# Aplicamos la funcin para a=3 y b=7 (con salida)
x = maximo(3, 7)
x
```

```r
# Comprobamos la estructura
str(x)
```

```r
# Hacemos otra prueba incluyendo una variable y una operacin aritmtica como argumento
y = 19
maximo(20, y + 7)
```

```r
# Otros ejemplo
x = maximo(2, 3)  # se asigna la salida de la funcin a x
x
```

```r
# Otro ejemplo
2 + maximo(4, 5.2) # la invocacin a la funcin forma parte de una expresin
```

**<u>Parmetros reales vs formales</u>.-** Tener en cuenta que al invocar a una funcin se evalan las expresiones que conforman los `parmetros reales` y se emparejan con los `parmetros formales`. Los **parmetros formales** son variables que reciben la informacin de entrada de la funcin (en el ejemplo, `a` y `b`). Los **parmetros reales** en el ejemplo seran `3` y `7`. Una vez emparejados los parmetros reales con los formales, se ejecutan las instrucciones de la funcin. Una `funcin devuelve` como salida el resultado de evaluar la ltima instruccin que ejecuta. En el ejemplo, la ltima instruccin ejecutada es `out`, que es una expresin que produce el mximo de los dos valores de entrada de la funcin (almacenados en a y b).

## 2.1 La funcin return

En el apartado anterior hemos comentado que `una funcin devuelve el resultado de evaluar la ltima instruccin que ejecuta`. Esto **es cierto si el cdigo de la funcin NO invoca a la funcin return**. La funcin return tiene un parmetro y produce el siguiente efecto: <u>termina la ejecucin de la funcin</u> en la que se ejecuta y el valor devuelto por la funcin es el parmetro con que es invocada `return`. 

**Ejemplo 4 <mark>[uso de return en funcin mximo]</mark>** La funcin maximo previa puede codificarse as incluyendo la funcin `return`.

```r
# Obtener el valor mximo de 2 valores (SIN funcin return)
maximo <- function(a, b) {
  if (a > b) {
    out <- a
  } else {
    out <- b
  }
  out
}
```

```r
# Aplicamos nuestra funcin a valores concretos de los parmetros
x = maximo(14, 8)
x
```

```r
str(x)
```

```r
# Obtener el valor mximo de 2 valores (CON funcin return)
maximo2 = function(a, b) {
  if (a > b) {
    out <- a
  } else {
    out <- b
  }
  return(out)
}
```

```r
# Aplicamos nuestra funcin a valores concretos de los parmetros
y = maximo2(14, 8)
y
```

```r
# Comprobamos la estructura
str(y)
```

```r
# Otra posibilidad es la siguiente
maximo3 = function(a, b) {
  if (a > b) {
    return(a)
  } else {
    return(b)
  }
}
```

```r
# Aplicamos la nueva funcin a valores concretos de los parmetros
z = maximo3(6, 7)
z
```

```r
# Comprobamos la estructura
str(z)
```

```r
# O incluso podramos definir la funcin del siguiente modo
maximo4 = function(a, b) {
  if (a > b) {
    return(a)
  }
  return(b)
}
```

```r
# Aplicamos la nueva funcin a valores concretos de los parmetros
w = maximo4(34, 19)
w
```

```r
# Aplicamos la nueva funcin a valores concretos de los parmetros
x = maximo(19, 34)
x
```

**<u>Utilidad de la funcin return</u>.-** En este ltimo ejemplo, la instruccin `return(b)` se ejecuta slamente si `a` no es mayor que `b`. Si `a` fuera mayor que `b` entonces se ejecutara `return(a)`, lo que **terminara la ejecucin de la funcin** e **impedira que se ejecutara** cualquier instruccin tras el `if`.

## 2.2 Las variables locales

Una funcin puede realizar clculos complejos, por lo que a veces **necesita usar variables para almacenar los resultados intermedios** de sus clculos.

**Ejemplo 5 <mark>[uso variables locales en funcin suma]</mark>** La siguiente funcin calcula la suma de los elementos de un vector.

```r
# Funcin que suma los elementos de un vector (suma acumulada)
suma = function(v) {
  s = 0
  for (x in v) {
    s = s + x
  }
  s
}
```

```r
# Aplicamos datos de entrada
suma(c(2, 5, -3))
```

**<u>Utilidad de las variables locales</u>.-** Las variables creadas en una funcin se llaman **variables locales**. En el ejemplo, la variable `s` es una variable local que se usa para acumular los elementos del vector. Las variables locales, junto con los parmetros formales, se crean al ejecutarse la funcin y normalmente desaparecen al terminar su ejecucin. **Las variables locales y parmetros formales son independientes de otras variables** que puedan existir fuera de la funcin con el mismo nombre como veremos en el siguiente ejemplo.

**Ejemplo 6 <mark>[uso parmetro formal y variable local]</mark>** Observar como el parmetro formal `v` y la variable local `s` coinciden en nombre con otras variables externas a la funcin. Al invocar a la funcin `suma` se crean las variables locales `v` y `s` y al terminar su ejecucin desaparecen, siendo independientes de las variables `v` y `s` que conservan sus valores.

```r
# Funcin que suma los elementos de un vector (suma acumulada) incluyendo variables previamente definidas
v = 5
s = 10
suma = function(v) {
  s = 0
  for (x in v) {
    s = s + x
  }
  s
}
```

```r
# Aplicamos la funcin sobre unos datos
suma(c(2, 5, -3))
```

```r
# Comprobamos que los valores de las variables v y s definidas no se han alterado
v; s
```

## 2.3 Las variables globales

Una funcin tambin puede **usar variables que se han creado fuera de la funcin** y estas se denominan **variables globales**.

**Ejemplo 7 <mark>[uso variables globales en funcin suma]</mark>** La funcin `f` usa el valor de la variable `w` que **NO es una variable local**, sino que es **global** a la funcin (<u>ha sido creada fuera</u>). Veamos cmo se utilizara con la funcin suma.

```r
# Definimos la funcin
w = 10
f = function() {
  w = w + 2
  w
}
```

## 2.5 Parmetros por defecto

Un **parmetro por defecto** es un parmetro formal que se define terminndolo con un igual y una expresin. Vamos a ver su utilidad con un ejemplo.

**Ejemplo 10 <mark>[uso de parmetros por defecto en funcin distancia eucldea]</mark>** La siguiente funcin sirve para calcular la distancia eucldea entre dos puntos en un espacio bidimensional y comparamos el uso SIN vs CON parmetros por defecto.

```r
# Distancia entre los puntos (x1, y1) y (x2, y2) - SIN parmetros por defecto
distancia_eu <- function(x1, y1, x2, y2) {
  sqrt((x1 - x2)^2 + (y1 - y2)^2)
}
```

```r
# Aplicamos a un caso particular
# Dando valores a TODOS los parmetros
distancia_eu(1, 1, 2, 2) # distancia entre (1, 1) y (2, 2)
```

```r
# Dando valores a PARTE de los parmetros 
distancia_eu(4, 4)       # distancia entre (4, 4) y ??
```

Como es muy comn calcular la distancia de un punto al origen (0,0), se decide que el segundo punto sea, **por defecto**, el origen.

```r
# Distancia entre los puntos (x1, y1) y (x2, y2) - CON parmetros por defecto
distancia_eu2 <- function(x1, y1, x2 = 0, y2 = 0) {
  sqrt((x1 - x2)^2 + (y1 - y2)^2)
}
```

```r
# Dando valores a TODOS los parmetros
distancia_eu2(1, 1, 2, 2) # distancia entre (1, 1) y (2, 2)
```

```r
# Dando valores a PARTE de los parmetros (utilizamos los valores por defecto)
distancia_eu2(4, 4)       # distancia entre (4, 4) y (0, 0)
```

**<u>Utilidad de los parmetros por defecto</u>.-** En esta ltima versin de la funcin con parmetros por defecto, cuando un punto sea el origen no es preciso especificarlo. El efecto de un parmetro por defecto es que, **si no se especifica al ser invocada la funcin, el parmetro formal toma el valor por defecto**, es decir, el especificado tras el operador igual en la definicin de la funcin.

# 3. La estructura y funcionamiento de los condicionales $\tt if$ y $\tt if else$ <a id="topic3"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Las **estructuras de control** establecen condicionales en nuestros cdigo por eso es importante que conozcamos como se usan las <mark>**expresiones de control ms tpicas**</mark>. Una expresin de control ser un **conjunto de rdenes** que nos permiten decidir por donde queremos que avance nuestro programa: qu hacemos si sucede A? Y si sucede B?

Por ejemplo, **qu condiciones deben cumplirse para realizar una operacin o qu debe ocurrir para ejecutar una funcin**. Esto es de gran <u>utilidad para determinar la lgica y el orden en que ocurren las operaciones, en especial al `definir funciones`</u>. Las estructuras de control ms usadas en R son las siguientes.

**<u>Utilidad de los condicionales $\tt if$ y $\tt if else$</u>.-** Las **estructuras de control condicionales** (o condicionales a secas) permiten que un programa `decida de manera automtica entre varias opciones en funcin de si se cumplen o no determinadas condiciones`. Estas estructuras tienen en R la misma estructura que en casi todos los otros lenguajes de programacin. Por un lado, tenemos la estructura:

```r
if(condicin){accin}
```

que indica a R que **si se satisface** la `condicin` especificada entre parntesis, **se lleve a cabo** la `accin` especificada entre llaves y luego se continua con el programa, **mientras que si no se satisface** dicha `condicin`, se continua con el programa **sin efectuar** la `accin`.

**Ejemplo 13a <mark>[uso de if en funcin cuadrado]</mark>** Si quisiramos definir una funcin que valiera $x^2$ si $x0$ tendramos que usar el siguiente condicional:

```r
# Definimos la funcin cuadrado con if
f=function(x){
  if(x<=0){x^2}
}
```

```r
# Aplicamos la funcin cuadrado con if a un par de ejemplos
f(-3)
f(3) # Qu ha pasado con este caso?
```

```r
if(condicin){accin1} else {accin2}
```

que indica a R que **si se satisface** la `condicin`, **se lleve a cabo** la `accin1` y **si no se satisface** dicha `condicin` **se lleve a cabo** la `accin2` y luego, en ambos casos, se continua con el programa.

**Ejemplo 13b <mark>[uso de if else en funcin cuadrado]</mark>** Si quisiramos definir una funcin que valiera $x^2$ si $x0$ y $x^3$ si $x0$, tendramos que usar el siguiente condicional:

```r
# Definimos la funcin cuadrado con if else
f=function(x){
  if(x<=0){x^2} else {x^3}
}
```

```r
# Aplicamos la funcin cuadrado con if else a un par de ejemplos
f(-3)
f(3)
```

**Ejemplo 14 <mark>[uso de if en funcin > 13 aos]</mark>** Vemaos otro ejemplo en el que si todas las edades son mayores de 13 aos hace un print con dicho texto.

```r
# Todas las edades son mayores de 13 aos
edades <- c(14, 24, 56, 31, 20, 87, 73)
if (all(edades >= 13)) { # Si todas las personas son mayores de 13 aos
  
  print("todos mayores de edad")
  
}
```

En caso de que **NO se cumplan** las condiciones (FALSE), **NO suceder nada**.

```r
# No todas las edades son mayores de 13 aos ==> NO hace nada
edades <- c(11, 24, 56, 31, 20, 87, 73)
if (all(edades >= 13)) { # Si todas las personas son mayores de edad
  
  print("todos mayores de edad")
  
}
```

Si tras el `if` tenemos adems un `else`, **cuando la condicin NO se cumpla**, se ejecutar el cdigo que haya tras dicha partcula, permitindonos decidir que sucede cuando **S se cumple y cuando NO se cumple**. 

**Ejemplo 15 <mark>[uso de if else en funcin > 18 aos - mayor de edad]</mark>** En el caso siguiente, las rdenes dentro del primer `if` se ejecutarn si todas las personas son mayores de edad. En el supuesto de <u>NO cumplirse (como es el caso)</u>, se ejecutar lo que hay entre llaves tras el `else` (<u>filtra solo los mayores de edad e imprime un mensaje por consola</u>).

```r
edades <- c(14, 24, 56, 31, 20, 87, 73)
if (all(edades >= 18)) { # Si todas las personas son mayores de edad
  
  print("todos mayores de edad")
  
} else { # si la condicin no se cumple: filtramos solo los mayores
  edades <- edades[edades >= 18]
  print("algn menor de edad se ha eliminado")
  
}
```

```r
# Vemos como el vector edades se ha sobreescrito solo con aquellas edades que cumplen la condicin
edades
```

**Ejemplo 16 <mark>[concatenacin de if else en funcin > 18 aos]</mark>** Dicha estructura puede **anidarse**, de forma que vayamos concatenando estructuras `ifelse`, como en la celda que tenemos debajo.

```r
edades <- c(14, 14, 16, 11, 2, 17, 13)
if (all(edades >= 18)) { # Si todas las personas son mayores de edad
  
  print("todos mayores de edad")
  
} else if (any(edades >= 18)) { # si alguna es mayor de edad
  edades_18 <- edades[edades >= 18]  
  print("algn menor de edad se ha eliminado")
  
} else { # ninguna persona mayor de edad
  
  print("todas las personas son menores de edad")
  
}
```

Esta **estructura condicional puede ser vectorizada**, de forma que podamos reunir en una **sola fila un nmero elevado de estructuras de comparacin** con la funcin `ifelse()`, cuyos argumentos de entrada sern la condicin a evaluar, lo que sucede cuando se cumple y lo que no, que aplicar a cada elemento del vector de entrada. 

**Ejemplo 17 <mark>[uso de ifelse en funcin sustitucin nombres]</mark>** Vamos a definir un vector de nombres con algunos ausentes (`NA`), y vamos a sustituir los ausentes por el texto `"nombre_desconocido"` (los que NO sean ausentes se quedan como estn).

```r
# Definimos el vector con los nombres
nombres <- c("Juan", "Mara", NA, NA, "Luca", "Carmen", "Javier",
             NA, "Carlos", NA, "Gregorio", "Paloma")
```

```r
# Aplicamos la funcin ifelse del siguiente modo:
# Si tiene ausente --> "nombre_desconocido"
# Si no tiene ausente --> nombres originales
nombres_na <- ifelse(is.na(nombres), "nombre_desconocido", nombres)
```

# 4. La estructura y funcionamiento de los bucles tipo $\tt for$  <a id="topic4"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Un bucle de tipo $\tt for$ es una estructura del tipo `Para todo  haz ` en la que **se repite una determinada operacin o secuencia de operaciones para todos los elementos de un vector** (tambin puede ser de una lista, pero para simplificar aqu siempre tomaremos un vector). Con R, este tipo de bucles se especifican con una instruccin de la forma:

```r
for(ndice in vector){accin}
```

Esta sintaxis indica a R que **para todo** `ndice` que **pertenezca** al `vector`, **efecte** la `accin` especificada entre las llaves. En esta instruccin el **ndice** juega un papel de **variable interna**, como las **variables entre los parntesis** en la definicin de funciones. 

**Ejemplo 18a <mark>[uso de for en funcin cuadrados]</mark>** Para construir el vector de los cuadrados de nmeros naturales del 1 al 10 por medio de un bucle de tipo $\tt for$ tendramos que escribir lo siguiente:

```r
# Definicin bucle tipo for
for(n in 1:10){
    Cuadrados[n]=n^2
}
```

<u>Muy importante.-</u> Nos da un **error** ya que antes hay que definir el vector de `Cuadrados` y darle un `valor inicial`, **aunque sea igualarlo a un vector vaco, para que luego R lo vaya construyendo**:

```r
# Iniciamos Cuadrados como un vector vaco y aplicamos el bucle tipo for
Cuadrados=c()
for(n in 1:10){
    Cuadrados[n]=n^2
}
```

```r
# Sacamos el resultado por pantalla
Cuadrados
```

**Ejemplo 18b <mark>[uso de for en funcin cuadrados alternativa]</mark>** En vez de empezar con el vector `Cuadrados` vaco, **podramos iniciarlo con su primera entrada igual a 1**. Y en vez de ir definiendo las entradas sucesivas de este vector en las diferentes iteraciones del bucle, tambin **podramos ir aadindolas al final**.

```r
# Iniciamos Cuadrados como un vector con datos y aplicamos el bucle for
Cuadrados2=c(1)  
for(n in 2:10){
    Cuadrados2=c(Cuadrados2,n^2)
}
```

```r
# Sacamos el resultado por pantalla
Cuadrados2
```

**<u>Optimizacin algortmica</u>.-** Desde el punto de vista de R, estos dos bucles de tipo $\tt for$ **tienen un defecto: como cada iteracin del bucle cambia la dimensin del vector `Cuadrados`, en cada iteracin R define un nuevo objeto `Cuadrados` en el que copia el contenido del `Cuadrados` anterior y le aade el nuevo elemento**. En <u>bucles ms complicados, esto podra ralentizar el clculo</u>.

**Ejemplo 19 <mark>[uso de for en funcin cuadrados con longitud conocida]</mark>** La `manera correcta de definir este bucle` en R es iniciar el vector `Cuadrados` como un vector de la longitud que va a tener al final, 10 en nuestro caso, y con el contenido que queramos, por ejemplo constante, y entonces ir modificando sus entradas una a una. Pero **esto NO siempre podemos aplicarlo ya que a veces no sabemos la longitud que va a tener el objeto**.

```r
# Iniciamos Cuadrados como un vector con longitud predefinida y aplicamos el bucle for
Cuadrados3=rep(1,10)  
for(n in 2:10){
    Cuadrados3[n]=n^2
}
```

```r
# Sacamos el resultado por pantalla
Cuadrados3
```

**Ejemplo 20 <mark>[uso de for en funcin vector nmeros]</mark>** Luego, un **bucle `for{}` es una estructura que nos permite ejecutar un conjunto de rdenes un nmero repetido de veces (finito y conocido)**: dado un conjunto de ndices, el bucle ir **recorriendo cada elemento** de dicho conjunto, y para cada uno de ellos ejecutar lo que tenga dentro de las llaves.

```r
variable <- NULL # vector donde guardaremos los pasos del bucle

for (i in 1:7) { 
  variable[i] <- i # R es silencioso: salvo que hagamos un print dentro del bucle no nos imprimir nada por pantalla
}
```

# 5. La estructura y funcionamiento de los bucles tipo $\tt while$  <a id="topic5"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Los bucles de tipo `while` **repiten una determinada secuencia de operaciones mientras una cierta condicin lgica se satisfaga** <u>sin conocer a priori el no. de repeticiones</u>. Su sintaxis es:

```r
while(condicin){accin}
```

e indica a R que **antes de efectuar por primera vez** la `accin` especificada entre las llaves y tras cada ejecucin de la misma, **compruebe si** la `condicin` especificada entre parntesis **se satisface**; en `caso afirmativo`, **vuelve a efectuar la accin**, y en `caso negativo`, **para la ejecucin** del bucle. 

**Ejemplo 24 <mark>[uso del bucle tipo while as for]</mark>** 

```r
# Iniciamos Cuadrados como un vector con datos y aplicamos el bucle for
Cuadrados=rep(1,10)
for(n in 1:10){
    Cuadrados[n]=n^2
}
```

```r
#Visualizamos el resultado
Cuadrados
```

Se podra reescribir como un condicional tipo `while` del siguiente modo:

```r
# Iniciamos un contador
n=1 
# Iniciamos el vector
Cuadrados2=c() 
# Mientras n sea menor o igual a 10...
while(n<=10){ 
  Cuadrados2[n]=n^2 #aadimos n^2 al vector de Cuadrados...
  n=n+1 #y aumentamos en 1 el contador n
}
```

```r
#Visualizamos el resultado
Cuadrados2
```

El funcionamiento paso a paso de este bucle `while`sera el siguiente:

+ Como $110$, definir $Cuadrados[1]=1$ y redefinir $n=2$;
+ Comprobar que $210$, y entonces definir $Cuadrados[2]=4$ y redefinir $n=3$;
+ Comprobar que $310$, y entonces definir $Cuadrados[3]=9$ y redefinir $n=4$;
+ 
+ Comprobar que $1010$, y entonces definir $Cuadrados[10]=100$ y redefinir $n=11$;
+ Comprobar que $11>10$, y parar.

**Ejemplo 25 <mark>[uso de while en la funcin Fibonacci]</mark>** Supongamos que queremos calcular los nmeros de Fibonacci menores o iguales a 100. Podemos usar un bucle de tipo $\tt while$ de la manera siguiente:

+ Iniciaremos un vector con los dos primeros nmeros de Fibonacci (1 y 1).
+ Mientras la suma de los dos ltimos nmeros de Fibonacci calculados sea menor o igual a 100, aadiremos esta suma al final del vector, como siguiente nmero de Fibonacci:

```r
# Iniciamos el vector de Fibonacci y aplicamos el bucle while
Fib=c(1,1) 
while(Fib[length(Fib)]+Fib[length(Fib)-1]<=100){
  Fib[length(Fib)+1]=Fib[length(Fib)]+Fib[length(Fib)-1] 
} 
```

```r
#Visualizamos el resultado
Fib
```

El siguiente nmero de Fibonacci ya sera mayor que 100 (verificamos que es correcto):

```r
Fib[length(Fib)]+Fib[length(Fib)-1]
```

**Ejemplo 26 <mark>[uso de while en funcin ciclos]</mark>** Luego como hemos visto otra manera de disear un bucle es con la **estructura `while{}`**, que ejecutar el **bucle un nmero de veces pero que a priori es desconocido** hasta que la condicin impuesta deje de ser `TRUE` (en el <u>`for` sabemos de antemano el nmero de iteraciones</u>).

```r
# Definimos las variables de entrada
max_ciclos = 10
ciclos = 1

# Mientras el nmero de ciclos sea inferior 10, imprime el siguiente texto
while(ciclos <= max_ciclos) {
  print(paste("Todava no, vamos por el ciclo ", ciclos)) # Pegamos la frase al nmero del ciclo con paste
  ciclos <- ciclos + 1
}
```

# 6. La estructura y funcionamiento de los comandos $\tt break/next/repeat$  <a id="topic6"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>

En `R` tenemos dos comandos reservados para poder **abortar un bucle** `break` o **avanzar forzosamente un bucle** `next`. La primera nos habilita para **parar un bucle** aunque no haya llegado al final de su conjunto de ndices a recorrer (o se siga cumpliendo la condicin del `while{}`).

**Ejemplo 27 <mark>[uso de break en funcin suprimir nmero]</mark>**  

```r
for(i in 1:10) {
  if (i == 7) {  
    break # si i es 7, el bucle frena aqu (nunca llegar a imprimir el 7 ni los sucesivos)  
  }
  print(i)
}
```

**Ejemplo 28 <mark>[uso de next en funcin suprimir nmero]</mark>** Este comando **obliga al bucle a avanzar a la siguiente interaccin**, abortando la iteracin actual en la que se encuentra. 

```r
for(i in 1:10) {
  if (i == 7) {  
    next # si i es 7, la iteracin frenar aqu y pasar a la siguiente por lo que imprimir todos menos el 7  
  }
  print(i)
}
```

**Ejemplo 29 <mark>[uso de repeat en funcin contar nmeros]</mark>** Aunque es una opcin **muy poco usada**, existe una estructura de control llamada `repeat{}` que nos **ejecuta un bucle de forma infinita** hasta que le ordenemos parar con un `break`.

```r
conteo <- 0
repeat {  
  conteo <- conteo + 1
  if (conteo >= 100) { break } 
}
```

```r
#Visualizamos el resultado
conteo
```

**<u>Nota</u>.-** Las estucturas `while{}` y `repeat{}` conllevan su riesgo a la hora de quedarnos en un loop permetuo, ya que si no incluimos un `break` o la condicin nunca llega a ser `TRUE`, el bucle seguir ejecutndose de forma infinita y solo podr ser detenido abortando la ejecucin con la tecla `ESC`.
