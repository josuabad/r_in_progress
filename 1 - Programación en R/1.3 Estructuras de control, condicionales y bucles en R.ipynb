# 1. Primeros pasos a la hora de programar una función en R <a id="topic1"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


El lenguaje R permite al usuario definir sus **propias funciones**. Una **función** permite escribir un bloque de código y ejecutarlo para distintos datos. Una función puede considerarse un subprograma que resuelve una subtarea. Un motivo para utilizar funciones es que permiten estructurar u organizar el código de un programa. Cuando hay que resolver un problema complejo, en lugar de intentar solucionarlo mediante un programa muy extenso es mejor **descomponerlo en subproblemas**.

En notebooks previos ya hemos usado algunas funciones de R, en este notebook, vamos a aprender a crear nuestras propias funciones. El esquema de una función es el que sigue:

```r
nombre_función = function(arg1, arg2, ... ){expresión}
```

+ En la expresión anterior `arg1, arg2, ...` son los argumentos de entrada (también llamados parámetros).

+ La `expresión` está compuesta de comandos que utilizan los argumentos de entrada para dar la **salida** deseada.

+ La **salida de una función** puese ser un `número, un vector, una grafica, un mensaje, etc`.

**Ejemplo 1 <mark>[función suma]</mark>** Vamos a empezar tratando de obtener una función que sume dos números. Por ejemplo el 3 y el 4. Si no creamos una función a cada paso tendríamos que ir utilizando el siguiente código **manualmente**.

```r
# Sumar números manualmente
3+4
```

Si queremos **automatizar** dicho proceso e **integrarlo en una función** tendríamos que operar de la siguiente manera. `Consideramos 2 números`, por tanto, `2 argumentos` a los que le vamos a llamar **'a'** y **'b'**. Teniendo en cuenta los principales elementos de una función descritos previamente tendríamos lo siguiente. Vamos a llamar a nuestra función `suma`.

```r
# Sumar números a través de una función creada por nosotros
suma = function(a,b){
    out = a+b
}
```

```r
# Aplicamos la función para a=3 y b=4
suma(3,4)
```

Se ha ejecutado **pero NO hemos obtenido ningún tipo de objeto de salida**. Para ello, deberíamos de haber incluido en nuestra función como siempre el propio objeto para que nos saque la información que contiene al ejecutarlo. 

```r
# Sumar números a través de una función creada por nosotros (con salida)
suma = function(a,b){
    out = a+b
    out
}
```

```r
# Aplicamos la función para a=3 y b=4 (con salida)
suma(3,4)
```

**<u>Descripción función suma</u>.-** Para crear una función **se usa la palabra reservada** `function`. A continuación se **enumeran los parámetros formales** de la función separados por comas y encerrados entre paréntesis (en el ejemplo los parámetros formales son `a` y `b`). Después viene el **código de la función**. Puesto que el código del ejemplo tiene más de una instrucción, hay que encerrarlo entre llaves. Con esto se crea un **objeto de salida** `out` donde se almacena la operación anterior. Finalmente, la función creada se asigna al objeto `suma`.

**Ejemplo 2 <mark>[función máximo]</mark>** Ahora vamos a crear una función que calcula el máximo de sus dos argumentos.

```r
# Obtener el valor máximo de 2 valores
maximo <- function(a, b) {
  if (a > b) {
    out <- a
  } else {
    out <- b
  }
  out
}
```

```r
# Aplicamos la función para a=3 y b=7 (con salida)
x = maximo(3, 7)
x
```

```r
# Comprobamos la estructura
str(x)
```

```r
# Hacemos otra prueba incluyendo una variable y una operación aritmética como argumento
y = 19
maximo(20, y + 7)
```

```r
# Otros ejemplo
x = maximo(2, 3)  # se asigna la salida de la función a x
x
```

```r
# Otro ejemplo
2 + maximo(4, 5.2) # la invocación a la función forma parte de una expresión
```

**<u>Parámetros reales vs formales</u>.-** Tener en cuenta que al invocar a una función se evalúan las expresiones que conforman los `parámetros reales` y se emparejan con los `parámetros formales`. Los **parámetros formales** son variables que reciben la información de entrada de la función (en el ejemplo, `a` y `b`). Los **parámetros reales** en el ejemplo serían `3` y `7`. Una vez emparejados los parámetros reales con los formales, se ejecutan las instrucciones de la función. Una `función devuelve` como salida el resultado de evaluar la última instrucción que ejecuta. En el ejemplo, la última instrucción ejecutada es `out`, que es una expresión que produce el máximo de los dos valores de entrada de la función (almacenados en a y b).

## 2.1 La función return

En el apartado anterior hemos comentado que `una función devuelve el resultado de evaluar la última instrucción que ejecuta`. Esto **es cierto si el código de la función NO invoca a la función return**. La función return tiene un parámetro y produce el siguiente efecto: <u>termina la ejecución de la función</u> en la que se ejecuta y el valor devuelto por la función es el parámetro con que es invocada `return`. 

**Ejemplo 4 <mark>[uso de return en función máximo]</mark>** La función maximo previa puede codificarse así incluyendo la función `return`.

```r
# Obtener el valor máximo de 2 valores (SIN función return)
maximo <- function(a, b) {
  if (a > b) {
    out <- a
  } else {
    out <- b
  }
  out
}
```

```r
# Aplicamos nuestra función a valores concretos de los parámetros
x = maximo(14, 8)
x
```

```r
str(x)
```

```r
# Obtener el valor máximo de 2 valores (CON función return)
maximo2 = function(a, b) {
  if (a > b) {
    out <- a
  } else {
    out <- b
  }
  return(out)
}
```

```r
# Aplicamos nuestra función a valores concretos de los parámetros
y = maximo2(14, 8)
y
```

```r
# Comprobamos la estructura
str(y)
```

```r
# Otra posibilidad es la siguiente
maximo3 = function(a, b) {
  if (a > b) {
    return(a)
  } else {
    return(b)
  }
}
```

```r
# Aplicamos la nueva función a valores concretos de los parámetros
z = maximo3(6, 7)
z
```

```r
# Comprobamos la estructura
str(z)
```

```r
# O incluso podríamos definir la función del siguiente modo
maximo4 = function(a, b) {
  if (a > b) {
    return(a)
  }
  return(b)
}
```

```r
# Aplicamos la nueva función a valores concretos de los parámetros
w = maximo4(34, 19)
w
```

```r
# Aplicamos la nueva función a valores concretos de los parámetros
x = maximo(19, 34)
x
```

**<u>Utilidad de la función return</u>.-** En este último ejemplo, la instrucción `return(b)` se ejecuta sólamente si `a` no es mayor que `b`. Si `a` fuera mayor que `b` entonces se ejecutaría `return(a)`, lo que **terminaría la ejecución de la función** e **impediría que se ejecutara** cualquier instrucción tras el `if`.

## 2.2 Las variables locales

Una función puede realizar cálculos complejos, por lo que a veces **necesita usar variables para almacenar los resultados intermedios** de sus cálculos.

**Ejemplo 5 <mark>[uso variables locales en función suma]</mark>** La siguiente función calcula la suma de los elementos de un vector.

```r
# Función que suma los elementos de un vector (suma acumulada)
suma = function(v) {
  s = 0
  for (x in v) {
    s = s + x
  }
  s
}
```

```r
# Aplicamos datos de entrada
suma(c(2, 5, -3))
```

**<u>Utilidad de las variables locales</u>.-** Las variables creadas en una función se llaman **variables locales**. En el ejemplo, la variable `s` es una variable local que se usa para acumular los elementos del vector. Las variables locales, junto con los parámetros formales, se crean al ejecutarse la función y normalmente desaparecen al terminar su ejecución. **Las variables locales y parámetros formales son independientes de otras variables** que puedan existir fuera de la función con el mismo nombre como veremos en el siguiente ejemplo.

**Ejemplo 6 <mark>[uso parámetro formal y variable local]</mark>** Observar como el parámetro formal `v` y la variable local `s` coinciden en nombre con otras variables externas a la función. Al invocar a la función `suma` se crean las variables locales `v` y `s` y al terminar su ejecución desaparecen, siendo independientes de las variables `v` y `s` que conservan sus valores.

```r
# Función que suma los elementos de un vector (suma acumulada) incluyendo variables previamente definidas
v = 5
s = 10
suma = function(v) {
  s = 0
  for (x in v) {
    s = s + x
  }
  s
}
```

```r
# Aplicamos la función sobre unos datos
suma(c(2, 5, -3))
```

```r
# Comprobamos que los valores de las variables v y s definidas no se han alterado
v; s
```

## 2.3 Las variables globales

Una función también puede **usar variables que se han creado fuera de la función** y estas se denominan **variables globales**.

**Ejemplo 7 <mark>[uso variables globales en función suma]</mark>** La función `f` usa el valor de la variable `w` que **NO es una variable local**, sino que es **global** a la función (<u>ha sido creada fuera</u>). Veamos cómo se utilizaría con la función suma.

```r
# Definimos la función
w = 10
f = function() {
  w = w + 2
  w
}
```

```r
# Aplicamos a un ejemplo
f()
```

## 2.4 Funciones que devuelven varios valores

<u>Las funciones que hemos visto hasta ahora **devuelven un escalar**</u>. No obstante, algunas funciones deben **devolver más de una valor**. Sin embargo, **las funciones de R sólo pueden devolver un objeto, resultado de evaluar una expresión**. Por lo tanto, cuando queramos devolver más de un dato habrá que **devolver un objeto que almacene una colección de datos**, como un vector, factor, matriz, lista o dataframe como vimos en el `Notebook 1.2`. Habrá que elegir el tipo de objeto que creamos que sea más adecuado a nuestros cálculos.

**Ejemplo 8 <mark>[función que devuelve > 1 salida - vector]</mark>** Vamos a replicar la función `range`que devuelve el mínimo y el máximo de los elementos de un vector.

```r
# Rango de los valores de un vector siendo el parámetro de entrada (v, el vector)
rango = function(v) {
  c(min(v), max(v))
}
```

```r
# Veamos un caso particular
set.seed(1)
v = sample(10, 4)
v
```

```r
# Aplicamos nuestra función
rango(v)
range(v) # para comparar si funciona igual que range
```

**Ejemplo 9 <mark>[función que devuelve > 1 salida - lista]</mark>** Hemos decidido devolver un `vector` con dos elementos: el mínimo y el máximo. Sin embargo, también podíamos haber optado por devolver una `lista` con el mínimo y el máximo. En general, las funciones que devuelven datos de distinto tipo usan una lista como valor de retorno, pues las listas permiten incluir datos de distintos tipos como vimos en el Notebook 1.2.

```r
# Rango de los valores de un vector
rango2 = function(v) {
  list(minimo = min(v), maximo = max(v))
}
```

```r
# Aplicamos nuestra función
rango2(v)
```

## 2.5 Parámetros por defecto

Un **parámetro por defecto** es un parámetro formal que se define terminándolo con un igual y una expresión. Vamos a ver su utilidad con un ejemplo.

**Ejemplo 10 <mark>[uso de parámetros por defecto en función distancia euclídea]</mark>** La siguiente función sirve para calcular la distancia euclídea entre dos puntos en un espacio bidimensional y comparamos el uso SIN vs CON parámetros por defecto.

```r
# Distancia entre los puntos (x1, y1) y (x2, y2) - SIN parámetros por defecto
distancia_eu <- function(x1, y1, x2, y2) {
  sqrt((x1 - x2)^2 + (y1 - y2)^2)
}
```

```r
# Aplicamos a un caso particular
# Dando valores a TODOS los parámetros
distancia_eu(1, 1, 2, 2) # distancia entre (1, 1) y (2, 2)
```

```r
# Dando valores a PARTE de los parámetros 
distancia_eu(4, 4)       # distancia entre (4, 4) y ??
```

Como es muy común calcular la distancia de un punto al origen (0,0), se decide que el segundo punto sea, **por defecto**, el origen.

```r
# Distancia entre los puntos (x1, y1) y (x2, y2) - CON parámetros por defecto
distancia_eu2 <- function(x1, y1, x2 = 0, y2 = 0) {
  sqrt((x1 - x2)^2 + (y1 - y2)^2)
}
```

```r
# Dando valores a TODOS los parámetros
distancia_eu2(1, 1, 2, 2) # distancia entre (1, 1) y (2, 2)
```

```r
# Dando valores a PARTE de los parámetros (utilizamos los valores por defecto)
distancia_eu2(4, 4)       # distancia entre (4, 4) y (0, 0)
```

**<u>Utilidad de los parámetros por defecto</u>.-** En esta última versión de la función con parámetros por defecto, cuando un punto sea el origen no es preciso especificarlo. El efecto de un parámetro por defecto es que, **si no se especifica al ser invocada la función, el parámetro formal toma el valor por defecto**, es decir, el especificado tras el operador igual en la definición de la función.

# 3. La estructura y funcionamiento de los condicionales $\tt if$ y $\tt if else$ <a id="topic3"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Las **estructuras de control** establecen condicionales en nuestros código por eso es importante que conozcamos como se usan las <mark>**expresiones de control más típicas**</mark>. Una expresión de control será un **conjunto de órdenes** que nos permiten decidir por donde queremos que avance nuestro programa: ¿qué hacemos si sucede A? ¿Y si sucede B?

Por ejemplo, **qué condiciones deben cumplirse para realizar una operación o qué debe ocurrir para ejecutar una función**. Esto es de gran <u>utilidad para determinar la lógica y el orden en que ocurren las operaciones, en especial al `definir funciones`</u>. Las estructuras de control más usadas en R son las siguientes.

**<u>Utilidad de los condicionales $\tt if$ y $\tt if else$</u>.-** Las **estructuras de control condicionales** (o condicionales a secas) permiten que un programa `decida de manera automática entre varias opciones en función de si se cumplen o no determinadas condiciones`. Estas estructuras tienen en R la misma estructura que en casi todos los otros lenguajes de programación. Por un lado, tenemos la estructura:

```r
if(condición){acción}
```

que indica a R que **si se satisface** la `condición` especificada entre paréntesis, **se lleve a cabo** la `acción` especificada entre llaves y luego se continua con el programa, **mientras que si no se satisface** dicha `condición`, se continua con el programa **sin efectuar** la `acción`.

**Ejemplo 13a <mark>[uso de if en función cuadrado]</mark>** Si quisiéramos definir una función que valiera $x^2$ si $x⩽0$ tendríamos que usar el siguiente condicional:

```r
# Definimos la función cuadrado con if
f=function(x){
  if(x<=0){x^2}
}
```

```r
# Aplicamos la función cuadrado con if a un par de ejemplos
f(-3)
f(3) # ¿Qué ha pasado con este caso?
```

```r
if(condición){acción1} else {acción2}
```

que indica a R que **si se satisface** la `condición`, **se lleve a cabo** la `acción1` y **si no se satisface** dicha `condición` **se lleve a cabo** la `acción2` y luego, en ambos casos, se continua con el programa.

**Ejemplo 13b <mark>[uso de if else en función cuadrado]</mark>** Si quisiéramos definir una función que valiera $x^2$ si $x⩽0$ y $x^3$ si $x⩾0$, tendríamos que usar el siguiente condicional:

```r
# Definimos la función cuadrado con if else
f=function(x){
  if(x<=0){x^2} else {x^3}
}
```

```r
# Aplicamos la función cuadrado con if else a un par de ejemplos
f(-3)
f(3)
```

**Ejemplo 14 <mark>[uso de if en función > 13 años]</mark>** Vemaos otro ejemplo en el que si todas las edades son mayores de 13 años hace un print con dicho texto.

```r
# Todas las edades son mayores de 13 años
edades <- c(14, 24, 56, 31, 20, 87, 73)
if (all(edades >= 13)) { # Si todas las personas son mayores de 13 años
  
  print("todos mayores de edad")
  
}
```

En caso de que **NO se cumplan** las condiciones (FALSE), **NO sucederá nada**.

```r
# No todas las edades son mayores de 13 años ==> NO hace nada
edades <- c(11, 24, 56, 31, 20, 87, 73)
if (all(edades >= 13)) { # Si todas las personas son mayores de edad
  
  print("todos mayores de edad")
  
}
```

Si tras el `if` tenemos además un `else`, **cuando la condición NO se cumpla**, se ejecutará el código que haya tras dicha partícula, permitiéndonos decidir que sucede cuando **SÍ se cumple y cuando NO se cumple**. 

**Ejemplo 15 <mark>[uso de if else en función > 18 años - mayor de edad]</mark>** En el caso siguiente, las órdenes dentro del primer `if` se ejecutarán si todas las personas son mayores de edad. En el supuesto de <u>NO cumplirse (como es el caso)</u>, se ejecutará lo que hay entre llaves tras el `else` (<u>filtra solo los mayores de edad e imprime un mensaje por consola</u>).

```r
edades <- c(14, 24, 56, 31, 20, 87, 73)
if (all(edades >= 18)) { # Si todas las personas son mayores de edad
  
  print("todos mayores de edad")
  
} else { # si la condición no se cumple: filtramos solo los mayores
  edades <- edades[edades >= 18]
  print("algún menor de edad se ha eliminado")
  
}
```

```r
# Vemos como el vector edades se ha sobreescrito solo con aquellas edades que cumplen la condición
edades
```

**Ejemplo 16 <mark>[concatenación de if else en función > 18 años]</mark>** Dicha estructura puede **anidarse**, de forma que vayamos concatenando estructuras `ifelse`, como en la celda que tenemos debajo.

```r
edades <- c(14, 14, 16, 11, 2, 17, 13)
if (all(edades >= 18)) { # Si todas las personas son mayores de edad
  
  print("todos mayores de edad")
  
} else if (any(edades >= 18)) { # si alguna es mayor de edad
  edades_18 <- edades[edades >= 18]  
  print("algún menor de edad se ha eliminado")
  
} else { # ninguna persona mayor de edad
  
  print("todas las personas son menores de edad")
  
}
```

Esta **estructura condicional puede ser vectorizada**, de forma que podamos reunir en una **sola fila un número elevado de estructuras de comparación** con la función `ifelse()`, cuyos argumentos de entrada serán la condición a evaluar, lo que sucede cuando se cumple y lo que no, que aplicará a cada elemento del vector de entrada. 

**Ejemplo 17 <mark>[uso de ifelse en función sustitución nombres]</mark>** Vamos a definir un vector de nombres con algunos ausentes (`NA`), y vamos a sustituir los ausentes por el texto `"nombre_desconocido"` (los que NO sean ausentes se quedan como están).

```r
# Definimos el vector con los nombres
nombres <- c("Juan", "María", NA, NA, "Lucía", "Carmen", "Javier",
             NA, "Carlos", NA, "Gregorio", "Paloma")
```

# 4. La estructura y funcionamiento de los bucles tipo $\tt for$  <a id="topic4"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Un bucle de tipo $\tt for$ es una estructura del tipo `Para todo … haz …` en la que **se repite una determinada operación o secuencia de operaciones para todos los elementos de un vector** (también puede ser de una lista, pero para simplificar aquí siempre tomaremos un vector). Con R, este tipo de bucles se especifican con una instrucción de la forma:

```r
for(índice in vector){acción}
```

Esta sintaxis indica a R que **para todo** `índice` que **pertenezca** al `vector`, **efectúe** la `acción` especificada entre las llaves. En esta instrucción el **índice** juega un papel de **variable interna**, como las **variables entre los paréntesis** en la definición de funciones. 

**Ejemplo 18a <mark>[uso de for en función cuadrados]</mark>** Para construir el vector de los cuadrados de números naturales del 1 al 10 por medio de un bucle de tipo $\tt for$ tendríamos que escribir lo siguiente:

```r
# Definición bucle tipo for
for(n in 1:10){
    Cuadrados[n]=n^2
}
```

<u>Muy importante.-</u> Nos da un **error** ya que antes hay que definir el vector de `Cuadrados` y darle un `valor inicial`, **aunque sea igualarlo a un vector vacío, para que luego R lo vaya construyendo**:

```r
# Iniciamos Cuadrados como un vector vacío y aplicamos el bucle tipo for
Cuadrados=c()
for(n in 1:10){
    Cuadrados[n]=n^2
}
```

```r
# Sacamos el resultado por pantalla
Cuadrados
```

**Ejemplo 18b <mark>[uso de for en función cuadrados alternativa]</mark>** En vez de empezar con el vector `Cuadrados` vacío, **podríamos iniciarlo con su primera entrada igual a 1**. Y en vez de ir definiendo las entradas sucesivas de este vector en las diferentes iteraciones del bucle, también **podríamos ir añadiéndolas al final**.

```r
# Iniciamos Cuadrados como un vector con datos y aplicamos el bucle for
Cuadrados2=c(1)  
for(n in 2:10){
    Cuadrados2=c(Cuadrados2,n^2)
}
```

```r
# Sacamos el resultado por pantalla
Cuadrados2
```

**<u>Optimización algorítmica</u>.-** Desde el punto de vista de R, estos dos bucles de tipo $\tt for$ **tienen un defecto: como cada iteración del bucle cambia la dimensión del vector `Cuadrados`, en cada iteración R define un nuevo objeto `Cuadrados` en el que copia el contenido del `Cuadrados` anterior y le añade el nuevo elemento**. En <u>bucles más complicados, esto podría ralentizar el cálculo</u>.

**Ejemplo 19 <mark>[uso de for en función cuadrados con longitud conocida]</mark>** La `manera correcta de definir este bucle` en R es iniciar el vector `Cuadrados` como un vector de la longitud que va a tener al final, 10 en nuestro caso, y con el contenido que queramos, por ejemplo constante, y entonces ir modificando sus entradas una a una. Pero **esto NO siempre podemos aplicarlo ya que a veces no sabemos la longitud que va a tener el objeto**.

```r
# Iniciamos Cuadrados como un vector con longitud predefinida y aplicamos el bucle for
Cuadrados3=rep(1,10)  
for(n in 2:10){
    Cuadrados3[n]=n^2
}
```

```r
# Sacamos el resultado por pantalla
Cuadrados3
```

**Ejemplo 20 <mark>[uso de for en función vector números]</mark>** Luego, un **bucle `for{}` es una estructura que nos permite ejecutar un conjunto de órdenes un número repetido de veces (finito y conocido)**: dado un conjunto de índices, el bucle irá **recorriendo cada elemento** de dicho conjunto, y para cada uno de ellos ejecutará lo que tenga dentro de las llaves.

```r
variable <- NULL # vector donde guardaremos los pasos del bucle

for (i in 1:7) { 
  variable[i] <- i # R es silencioso: salvo que hagamos un print dentro del bucle no nos imprimirá nada por pantalla
}
```

```r
# Sacamos el resultado por pantalla
variable
```

**Ejemplo 21 <mark>[uso de for en función print números]</mark>** Lo que tenemos dentro de los paréntesis, en la línea del `for`, no es más que la secuencia de números que hemos aprendido a construir. No siempre tenemos que realizar una asignación o cambiar una variable dentro de un bucle, **podemos ejecutar acciones directamente sin asignar o guardar nada**.

```r
for (i in 1:7) { 
  print(i^3) # imprimimos cada índice al cubo
}
```

**Ejemplo 22 <mark>[uso de for en función dias de la semana]</mark>** Aunque normalmente el conjunto que recorre el bucle suelen ser índices numéricos, **podemos recorrer cualquier tipo de objeto**.

```r
dias_semana <- c("lunes", "martes", "miércoles", "jueves",
                 "viernes", "sábado", "domingo")
nombre_mayuscula <- NULL

for (dias in dias_semana) { # dias recorre los días de la semana tomando sus valores
  print(toupper(dias)) # Imprimimos en mayúsculas.
}
```

**Ejemplo 23 <mark>[comparación de for vs apply]</mark>** En muchos casos **es posible reemplazar** un bucle ${\tt for()}$ por un ${\tt apply()}$ igual que ocurre con otras estructuras de control. De hecho, ${\tt apply()}$ está **haciendo lo mismo** que un ${\tt for()}$, está iterando una operación en todos los elementos de una estructura de datos. Veamos ahora un ejemplo de lo comentado. 

<mark>**Todas las funciones de esta familia tienen una característica en común reciben como argumentos a un objeto y al menos una función**</mark>. 

Hasta ahora, todas las funciones que hemos usado han recibido como argumentos estructuras de datos, sean vectores, dataframes o matrices, por ejemplo. Las funciones de la familia ${\tt apply}$ tienen la particularidad que pueden recibir a otra función como un argumento.

La familia ${\tt apply}$ está formada por las siguientes 8 funciones (ordenadas alfabéticamente):

+ ${\tt mapply()}$

+ ${\tt sapply()}$
+ ${\tt tapply()}$

Nosotros repasaremos la función más general y de uso común de esta familia: la función ${\tt apply()}$. Esta función nos permitirá solucionar casi todos los problemas que nos encontremos. Además, conociendo su uso, las demás funciones de la familia ${\tt apply}$ serán relativamente fáciles de entender. **La sintaxis de esta función es**:

```r
# apply(X, MARGIN, FUN, ...)
```

+ `X`: una matriz o cualquier elemento que podamos convertir a una matriz, es decir, un dataframe o un vector normalmente.
+ `MARGIN`: la dimensión que agrupará los elementos de la matriz `X`, para aplicarles una función. Son identificadas con números, 1 son filas y 2 son colummnas.

+ `...`: argumentos opcionales que serán usados por `FUN`.

# 5. La estructura y funcionamiento de los bucles tipo $\tt while$  <a id="topic5"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Los bucles de tipo `while` **repiten una determinada secuencia de operaciones mientras una cierta condición lógica se satisfaga** <u>sin conocer a priori el no. de repeticiones</u>. Su sintaxis es:

```r
while(condición){acción}
```

e indica a R que **antes de efectuar por primera vez** la `acción` especificada entre las llaves y tras cada ejecución de la misma, **compruebe si** la `condición` especificada entre paréntesis **se satisface**; en `caso afirmativo`, **vuelve a efectuar la acción**, y en `caso negativo`, **para la ejecución** del bucle. 

**Ejemplo 24 <mark>[uso del bucle tipo while as for]</mark>** 

```r
# Iniciamos Cuadrados como un vector con datos y aplicamos el bucle for
Cuadrados=rep(1,10)
for(n in 1:10){
    Cuadrados[n]=n^2
}
```

```r
# Visualizamos el resultado
Cuadrados
```

Se podría reescribir como un condicional tipo `while` del siguiente modo:

```r
# Iniciamos un contador
n=1 
# Iniciamos el vector
Cuadrados2=c() 
# Mientras n sea menor o igual a 10...
while(n<=10){ 
  Cuadrados2[n]=n^2 #añadimos n^2 al vector de Cuadrados...
  n=n+1 #y aumentamos en 1 el contador n
}
```

```r
# Visualizamos el resultado
Cuadrados2
```

El funcionamiento paso a paso de este bucle `while`sería el siguiente:

+ Como $1⩽10$, definirá $Cuadrados[1]=1$ y redefinirá $n=2$;
+ Comprobará que $2⩽10$, y entonces definirá $Cuadrados[2]=4$ y redefinirá $n=3$;
+ Comprobará que $3⩽10$, y entonces definirá $Cuadrados[3]=9$ y redefinirá $n=4$;
+ …
+ Comprobará que $10⩽10$, y entonces definirá $Cuadrados[10]=100$ y redefinirá $n=11$;
+ Comprobará que $11>10$, y parará.

**Ejemplo 25 <mark>[uso de while en la función Fibonacci]</mark>** Supongamos que queremos calcular los números de Fibonacci menores o iguales a 100. Podemos usar un bucle de tipo $\tt while$ de la manera siguiente:

+ Iniciaremos un vector con los dos primeros números de Fibonacci (1 y 1).
+ Mientras la suma de los dos últimos números de Fibonacci calculados sea menor o igual a 100, añadiremos esta suma al final del vector, como siguiente número de Fibonacci:

```r
# Iniciamos el vector de Fibonacci y aplicamos el bucle while
Fib=c(1,1) 
while(Fib[length(Fib)]+Fib[length(Fib)-1]<=100){
  Fib[length(Fib)+1]=Fib[length(Fib)]+Fib[length(Fib)-1] 
} 
```

```r
# Visualizamos el resultado
Fib
```

El siguiente número de Fibonacci ya sería mayor que 100 (verificamos que es correcto):

```r
Fib[length(Fib)]+Fib[length(Fib)-1]
```

**Ejemplo 26 <mark>[uso de while en función ciclos]</mark>** Luego como hemos visto otra manera de diseñar un bucle es con la **estructura `while{}`**, que ejecutará el **bucle un número de veces pero que a priori es desconocido** hasta que la condición impuesta deje de ser `TRUE` (en el <u>`for` sabemos de antemano el número de iteraciones</u>).

```r
# Definimos las variables de entrada
max_ciclos = 10
ciclos = 1

# Mientras el número de ciclos sea inferior 10, imprime el siguiente texto
while(ciclos <= max_ciclos) {
  print(paste("Todavía no, vamos por el ciclo ", ciclos)) # Pegamos la frase al número del ciclo con paste
  ciclos <- ciclos + 1
}
```

# 6. La estructura y funcionamiento de los comandos $\tt break/next/repeat$  <a id="topic6"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>

En `R` tenemos dos comandos reservados para poder **abortar un bucle** `break` o **avanzar forzosamente un bucle** `next`. La primera nos habilita para **parar un bucle** aunque no haya llegado al final de su conjunto de índices a recorrer (o se siga cumpliendo la condición del `while{}`).

**Ejemplo 27 <mark>[uso de break en función suprimir número]</mark>**  

```r
for(i in 1:10) {
  if (i == 7) {  
    break # si i es 7, el bucle frena aquí (nunca llegará a imprimir el 7 ni los sucesivos)  
  }
  print(i)
}
```

**Ejemplo 28 <mark>[uso de next en función suprimir número]</mark>** Este comando **obliga al bucle a avanzar a la siguiente interacción**, abortando la iteración actual en la que se encuentra. 

```r
for(i in 1:10) {
  if (i == 7) {  
    next # si i es 7, la iteración frenará aquí y pasará a la siguiente por lo que imprimirá todos menos el 7  
  }
  print(i)
}
```

**Ejemplo 29 <mark>[uso de repeat en función contar números]</mark>** Aunque es una opción **muy poco usada**, existe una estructura de control llamada `repeat{}` que nos **ejecuta un bucle de forma infinita** hasta que le ordenemos parar con un `break`.

```r
conteo <- 0
repeat {  
  conteo <- conteo + 1
  if (conteo >= 100) { break } 
}
```

```r
# Visualizamos el resultado
conteo
```

**<u>Nota</u>.-** Las estucturas `while{}` y `repeat{}` conllevan su riesgo a la hora de quedarnos en un loop permetuo, ya que si no incluimos un `break` o la condición nunca llega a ser `TRUE`, el bucle seguirá ejecutándose de forma infinita y solo podrá ser detenido abortando la ejecución con la tecla `ESC`.
