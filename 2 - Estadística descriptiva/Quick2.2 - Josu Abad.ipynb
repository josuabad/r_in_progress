<div style="background-color:#0BF7CC"><b><u>Quick 1</u>. Supongamos que tenemos una variable aleatoria $Y$ que sigue una distribución uniforme continua en el intervalo $[-2, 3]$. Calcular la función de masa de probabilidad, la función de distribución acumulativa y representa ambas gráficamente. Interpretar los resultados y explicar cada uno de los argumentos utilizados en las funciones de masa y probabilidad. </b></div>

```r
rango <- c(-2, 3)
rango
```

```r
y <- seq(rango[1], rango[2], by = 0.01)
y
```

Hay que generar la secuencia de números desde el mínimo y el máximo solicitado, en este caso -2 y 3

```r
masa_probabilidad <- dunif(x = y, min = rango[1], max = rango[2])
masa_probabilidad
```

Como no se repiten números, la probabilidad de que cada uno de los que tenemos en esta secuencia "Y" tienen la misma probabilidad de aparecer que es 0.2

```r
plot(masa_probabilidad)
```

```r
distribucion_acumulativa <- punif(q = y, min = rango[1], max = rango[2])
distribucion_acumulativa
```

Con la secuencia de números obtenida nos damos cuenta que cada uno se separa del anterior por 0.002, lo que nos demuestra que el crecimiento es lineal y constante. Nos debería de dar una gráfica que represente esta situación (crecimiento lineal)

```r
plot(x = y,
     y = distribucion_acumulativa,
     type = "l",
     main = "Distribución acumulativa",
     xlab = "Secuencia numérica \"Y\"",
     ylab = "Distribución acumulativa"
    )
```

<div style="background-color:#0BF7CC"><b><u>Quick 4</u>. Supongamos que en una población de 500 personas, 300 son mujeres y 200 son hombres. Se selecciona una muestra aleatoria de 50 personas. Calcular la probabilidad de que al menos 30 de ellas sean mujeres y representar gráficamente la función de masa. </b></div>

```r
# Parámetros de la distribución hipergeométrica
N <- 500  # Tamaño total de la población (300 mujeres + 200 hombres)
m <- 300  # Número de mujeres en la población
n <- N-m  # Número de personas que no son mujeres en la población
k <- 50   # Tamaño de la muestra
```

```r
# Valor de interés (al menos 30 mujeres en la muestra)
q <- 30:n
```

```r
piper <- dhyper(q, m, n, k)
piper
```

```r
plot(piper, type = "l")
```
