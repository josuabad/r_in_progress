# 1. Introducción al análisis exploratorio de datos <a id="topic1"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Cuando contamos con una base de datos que describe aspectos de cualquier tipo de fenómeno físico, social, etc. y queremos llevar a cabo un *análisis estadístico* podemos optar por *dos enfoques diferentes*.

+ **Análisis exploratorio o descriptivo**: si nuestro objetivo es resumir, representar y explicar los datos concretos de los que disponemos. La estadística descriptiva es el conjunto de técnicas que se usan con este fin.
+ **Análisis inferencial**: si nuestro objetivo es deducir (inferir), a partir de estos datos, información significativa sobre el total de la población o las poblaciones de interés. Las técnicas que se usan en este caso forman la estadística inferencial.

Existe relación entre ambos. Cualquier análisis inferencial se suele empezar explorando los datos que se usarán así cómo también muchas técnicas descriptivas permiten estimar propiedades de la población de la que se ha extraído la muestra. En este caso, nos centraremos en entender algunas técnicas básicas de **estadística descriptiva** orientadas al análisis de datos. Estas consistirán en una serie de medidas, gráficos y estadísticos descriptivos que nos permitirán resumir y explorar un conjunto de datos. Un resumen de ellas aparece en la siguiente imagen.

# 2. ¿Cuáles son las principales tablas y gráficos de frecuencias? <a id="topic2"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>

## 2.1 Tablas de frecuencias unidimensionales

Una <u>tabla de frecuencias</u> muestra de forma ordenada un conjunto de datos estadísticos y a cada uno de ellos le asigna una frecuencia que, en pocas palabras, son las **veces que se repite un número o dato**. Puedes usar las tablas de frecuencias para ordenar variables cuantitativas o cualitativas. Existen diferentes tipos de tablas de frecuencias:

1. **Frecuencias absolutas**: son el número de veces que se repite un número en un conjunto de datos.
2. **Frecuencias relativas**: corresponden a las veces que se repite un número en un conjunto de datos respecto al total y se expresa en porcentajes (%).

**Ejemplo 1 <mark>[frecuencia absoluta]</mark>** Supongamos que tenemos una variable cuantitativa o cualitativa guardada en un vector o factor, veamos cómo calculamos la frecuencia absoluta y su interpretación. 

```r
# Generamos el vector de números
x=c(3,2,5,1,3,1,5,6,2,2,2,1,3,5,2)
x
```

```r
# Generamos el factor con las respuestas Si/No
Respuestas=factor(c("No","No","Sí","No","Sí","No","No","Sí"))
Respuestas
```

La **tabla de frecuencias absolutas** de un vector o factor que representa una variable cuantitativa o cualitativa se calcula con la función `table`. 

```r
# Obtenemos la tabla de frecuancias absolutas del vector de números
table(x)
```

```r
# Obtenemos la tabla de frecuancias absolutas del factor de respuestas
table(Respuestas)
```

**<u>Objeto creado por la función table</u>.-** El resultado de una función table es un objeto de datos de un tipo nuevo: una *tabla de contingencia*. Como vemos, al aplicar table a un vector obtenemos una tabla unidimensional formada por una fila con los niveles de la variable y una segunda fila donde, debajo de cada nivel, aparece su frecuencia absoluta en el vector.

```r
# Los nombres de las columnas de una tabla unidimensional se obtienen con la función names()
names(table(x))
```

```r
# Los nombres de las columnas de una tabla unidimensional se obtienen con la función names()
names(table(Respuestas))
```

**<u>Nombres de niveles ocultos</u>.-** Habréis observado que tras aplicar la función table a un vector sólo aparecen los nombres de los niveles presentes en el vector. Si el tipo de datos cualitativos usado tenía más **niveles** y queremos que **aparezcan explícitamente** en la tabla (con frecuencia $0$), hay que *transformar* el *vector* en un *factor* con los niveles deseados.

```r
# Creamos un nuevo factor y le asignamos los niveles que queremos tener en cuenta
z=factor(x, levels=1:7)  #Los niveles serán 1,2,3,4,5,6,7
z
```

```r
# Obtenemos la tabla de frecuancias absolutas del factor de números creado
table(z)
```

**Ejemplo 2 <mark>[acceder a elementos de una table]</mark>** A efectos prácticos, podemos pensar que una tabla unidimensional es como un vector de números donde cada entrada está identificada por un nombre: el de su columna. Para referirnos a una entrada de una tabla unidimensional, podemos usar tanto su posición como su nombre (entre comillas, aunque sea un número). Veamos algunos ejemplos. 

```r
# La cuarta columna de table(z)
table(z)[4]
# La columna de table(z) correspondiente al nivel 5:
table(z)["5"]
# El triple de la segunda columna de table(z)
3*table(z)[2]
```

**Ejemplo 3 <mark>[frecuencia relativa]</mark>** La *tabla de frecuencias relativas* de un vector se puede calcular aplicando la función `prop.table` a su table. El resultado vuelve a ser una tabla de contingencia unidimensional.

```r
# Obtenemos la tabla de frecuancias relativas del vector de números
prop.table(table(x))
```

```r
# Obtenemos la tabla de frecuancias relativas del factor de respuestas
prop.table(table(Respuestas))
```

**<u>Importante</u>.-** Tener en cuenta que la función prop.table se tiene que **aplicar al resultado de table**, NO al vector original. 

Si aplicamos prop.table a un vector de palabras o a un factor, **dará un error**, pero si la aplicamos a un vector de números, nos dará una tabla. Esta tabla NO es la tabla de frecuencias relativas de la variable cualitativa representada por el vector, sino la de una variable que tuviera como tabla de frecuencias absolutas este vector de números, entendiendo que cada entrada del vector representa la frecuencia de un nivel diferente. 

```r
# Aplicación errónea de la función prop.table() que no es la tabla de frecuencias relativas de x
prop.table(x)
```

**Ejemplo 4 <mark>[frecuencia relativa utilizando table]</mark>** También podemos calcular la tabla de frecuencias relativas de un vector dividiendo el resultado de table por el número de observaciones del siguiente modo.

```r
# Obtenemos la tabla de frecuancias relativas del vector de números operando con la función table()
table(x)/length(x)
table(x)/sum(table(x)) # Otra alternativa
```

**Ejemplo 5 <mark>[moda utilizando frecuencia absoluta]</mark>** Dados un vector x y un número natural n, la instrucción `names(which(table(x)==n)` nos da los niveles de x que tienen frecuencia absoluta igual a *n*.

```r
# Obtenemos la frecuencia absoluta del vector de números que sea igual a n operando con la función table() y which()
table(x)
names(which(table(x)==3)) # cuando n = 3
```

```r
# Obtenemos la moda del vector de números operando con la función table() y which()
names(which(table(x)==max(table(x))))
names(which(table(Respuestas)==max(table(Respuestas))))
```

## 2.2 Tablas de frecuencias bidimensionales

La función table también permite construir **tablas de frecuencias conjuntas de dos o más variables**. 

**Ejemplo 6 <mark>[frecuencia absoluta para 2 variables]</mark>** Supongamos que el vector `Respuestas`, de la sección anterior, contiene las respuestas a una pregunta dadas por unos individuos cuyos sexos tenemos almacenados en un vector `Sexo`, en el mismo orden que sus respuestas. En este caso, podemos construir una tabla que nos diga cuántos individuos de cada sexo han dado cada respuesta.

```r
# Obtenemos la tabla de frecuancias absolutas de las dos variables Respuestas y Sexo
Respuestas=c("No","No","Sí","No","Sí","No","No","Sí")
Sexo=c("M","M","M","H","H","H","H","H") # H es hombre, M es mujer
table(Respuestas,Sexo)
```

```r
# Modificamos el orden de las filas y columnas
table(Sexo,Respuestas)
```

**<u>Transponer filas y columnas</u>.-** Como podemos ver, en una tabla bidimensional producida aplicando table a dos vectores, los niveles del primer vector en el argumento definen las filas, y los del segundo, las columnas. Así, en table(Respuestas,Sexo), las filas corresponden a las respuestas y las columnas a los sexos. Para intercambiar filas por columnas, es decir, para “trasponer” la tabla sin tener que recalcularla, podemos usar la misma función `t()` que usamos para trasponer matrices:

```r
# Transponemos el orden de las filas y columnas 
t(table(Respuestas,Sexo))
```

**Ejemplo 7 <mark>[acceder a elementos de una table bidimensional]</mark>** Para referirnos a una entrada de una tabla bidimensional podemos usar corchetes como si estuviéramos en una matriz o un data frame. Dentro de los corchetes, tanto podemos usar los índices como los nombres (entre comillas) de los niveles.

```r
# El elemento de la 1º fila y la 2º columna de table(Respuestas,Sexo)
table(Respuestas,Sexo)[1,2]

# El elemento de la fila con valor igual a 'No' y la columna 'M' de table(Respuestas,Sexo)
table(Respuestas,Sexo)["No","M"]
```

**Ejemplo 8 <mark>[frecuencias relativas bidimensionales: globales vs marginales]</mark>** Como en el caso unidimensional, la función `prop.table` sirve para calcular tablas bidimensionales de frecuencias relativas conjuntas de pares de variables. Pero en el caso bidimensional tenemos dos tipos de frecuencias relativas, que definen, para cada par de variables, tres tablas diferentes:

+ Las frecuencias relativas **globales**: para cada par de niveles, uno de cada variable, la <u>fracción de individuos que pertenecen a ambos niveles respecto del total de la muestra</u>; por ejemplo, la fracción de mujeres que han contestado que sí respecto del total de la muestra sería la frecuencia relativa global del par (mujer, sí). La tabla de frecuencias relativas globales se calcula aplicando sin más la función `prop.table` a la table.


+ Las frecuencias relativas **marginales**: dentro de cada nivel de una variable, y para cada nivel de la otra, la <u>fracción de individuos que pertenecen al segundo nivel respecto del total de la subpoblación definida por el primer nivel</u>; por ejemplo, la fracción de mujeres que han contestado que sí respecto del total de mujeres sería una frecuencia relativa marginal. Dadas dos variables, <u>se pueden calcular dos familias de frecuencias relativas marginales</u>, según cuál sea la variable que defina las subpoblaciones en las que calculemos las frecuencias relativas de los niveles de la otra variable. La variable que define las subpoblaciones en las que las calculamos se indica con el parámetro `margin`. Es decir, con `margin=1` especificamos que la variable que define las subpoblaciones es la primera, y que, por lo tanto, las frecuencias relativas se calculan dentro de las filas; en cambio, con `margin=2` especificamos que la variable que define las subpoblaciones es la segunda, por lo que las frecuencias relativas se calculan por columnas; `margin=NULL` es su valor por defecto y, por lo tanto, equivalente a no especificar este parámetro.

**<u>Interpretación resultados</u>.-** De esta manera, por ejemplo, en la primera tabla podemos leer que un 25% del total de la muestra son mujeres que han contestado que no; en la segunda tabla, que un 66.67% de las mujeres han contestado que no; y en la tercera tabla, que las mujeres representan el 40% del total de las personas que han contestado que no.

```r
# La tabla prop.table(table(Sexo,Respuestas)) nos da la fracción del total que representa 
# cada pareja (sexo, respuesta)
prop.table(table(Sexo,Respuestas))
```

```r
# La tabla prop.table(table(Sexo,Respuestas), margin=1) nos da la fracción que representa 
# cada respuesta dentro de cada sexo
prop.table(table(Sexo,Respuestas), margin=1)
```

```r
# La tabla prop.table(table(Sexo,Respuestas), margin=2) nos da la fracción que representa 
# cada sexo dentro de cada respuesta
prop.table(table(Sexo,Respuestas), margin=2)
```

**Ejemplo 9 <mark>[todas las frecuencias relativas en una tabla]</mark>** La función `CrossTable` de la librería $\tt gmodels$ permite producir (especificando el parámetro `prop.chisq=FALSE`) un resumen de la tabla de frecuencias absolutas y las tres tablas de frecuencias relativas de dos variables en un formato adecuado para su visualización.

```r
# Instalamos la librería gmodels
install.packages("gmodels")
# Cargamos la librería gmodels
library(gmodels) 
# Obtenemos los resultados de todas las frecuencias relativas en una tabla
CrossTable(Sexo, Respuestas, prop.chisq=FALSE)
```

**<u>Descripción función CrossTable</u>.-** La leyenda Cell Contents explica los contenidos de cada celda de la tabla: en este caso, y en orden descendente, la frecuencia absoluta N, la frecuencia relativa por filas, la frecuencia relativa por columnas, y la frecuencia relativa global. Asimismo, se muestran las celdas de los márgenes, con las frecuencias absolutas y relativas de cada fila (en la columna Row Total) y cada columna (en la fila Column Total). Esta función dispone de muchos parámetros que permiten modificar el contenido de las celdas y que podéis consultar en su Ayuda.

```r
# Ayuda de la función CrossTable
help(CrossTable)
```

## 2.3 Tablas de frecuencias multidimensionales

En general, podemos calcular **tablas de frecuencias de cualquier número de variables**, no sólo de una o dos. El manejo de estas tablas multidimensionales es similar al caso bidimensional, simplemente recordando que ahora hay más variables que tener en cuenta en el momento, por ejemplo, de especificar entradas o de calcular frecuencias relativas marginales. 

**Ejemplo 10 <mark>[frecuencia absoluta para 3 variables]</mark>** Supongamos que, además de los vectores utilizados de `Respuestas`y `Sexo`, tenemos un tercer vector con las nacionalidades de los individuos `Pais`.

```r
# Leemos las variables que vamos a utilizar
Respuestas=c("No","No","Sí","No","Sí","No","No","Sí")
Sexo=c("M","M","M","H","H","H","H","H")
Pais=c("Francia","Alemania","Italia","Italia","Italia","Italia","Alemania","Francia")
```

```r
# Podemos calcular entonces una tabla de frecuencias absolutas para las ternas (sexo, respuesta, país).
table(Sexo,Respuestas,Pais)
```

**<u>Descripción función table caso multidimensional</u>.-** R muestra la tabla tridimensional que obtenemos como una lista de tablas bidimensionales `table(Sexo,Respuestas)`, separando la población según los niveles de la tercera variable. Si no os gusta esta manera de visualizar una tabla tridimensional, una alternativa es usar la función `ftable`, que la mostrará en lo que se llama formato plano:

```r
# Visualización compacta de una tabla tridimensional
ftable(Sexo,Respuestas,Pais)
```

```r
# Los parámetros row.vars y col.vars de ftable permiten especificar qué variables queremos que aparezcan 
# como filas o como columnas, respectivamente
ftable(Sexo,Respuestas,Pais, col.vars=c("Sexo","Respuestas"))
```

```r
# Para referirnos a una entrada, o a una subtabla, de una tabla podemos usar corchetes
table(Sexo,Respuestas,Pais)["H","Sí","Italia"] # Hombre, Si, Italianos
```

```r
table(Sexo,Respuestas,Pais)[ , ,"Italia"] # Italianos
```

```r
table(Sexo,Respuestas,Pais)[  ,"Sí","Italia"] # Si, Italianos
```

```r
table(Sexo,Respuestas,Pais)["M", ,"Italia"] # Mujer, Italianos
```

**Ejemplo 10 <mark>[frecuencias relativas para 3 variables]</mark>** En una tabla multidimensional, podemos calcular frecuencias relativas marginales respecto de los niveles de una variable o respecto de combinaciones de niveles de varias variables. Por ejemplo, <u>las frecuencias relativas marginales de las respuestas en cada combinación (sexo, país)</u>. 

Como en el caso bidimensional, las tablas correspondientes se calculan aplicando `prop.table` a la tabla de frecuencias absolutas, y especificando con el parámetro `margin` las dimensiones, o combinaciones de dimensiones, respecto de las que calculamos las frecuencias relativas. Si no se especifica el parámetro margin, se obtiene la tabla de frecuencias relativas globales.

```r
# La tabla prop.table(table(Sexo,Respuestas,Pais)) nos da la fracción que representa 
# cada terna (sexo, respuesta, país) dentro del total de la muestra:
prop.table(table(Sexo,Respuestas,Pais))
```

```r
# La tabla prop.table(table(Sexo,Respuestas,Pais), margin=3) nos da la fracción que representa 
# cada pareja (sexo, respuesta) dentro de cada país:
prop.table(table(Sexo,Respuestas,Pais), margin=3)
```

**<u>Interpretación resultados</u>.-** En la primera tabla podemos leer que los hombres italianos que han contestado afirmativamente forman un 12.5% del total de individuos; en la segunda, que los hombres que han contestado que sí representan un 25% del total de individuos italianos; y en la tercera, que un 33.33% del total de hombres españoles ha contestado que sí.

```r
# Al aplicar prop.table al resultado de una ftable, la tabla resultante sigue en formato plano:
prop.table(ftable(Sexo,Respuestas,Pais)) 
```

**Ejemplo 11 <mark>[apply con la función table]</mark>** Podemos aplicar funciones a filas o columnas de una tabla multidimensional con la función `apply`, especificando en el parámetro $\tt FUN$ la función y en el parámetro $\tt  MARGIN$ la variable o combinación de variables a la que aplicamos la función (son las variables que aparecerán en la tabla resultante). Observar las dos instrucciones siguientes y sus resultados:

```r
# Sumamos los números de respuestas agrupando por combinaciones (sexo,respuesta):
apply(table(Sexo,Respuestas,Pais), MARGIN=c(1,2), FUN=sum)
```

```r
# Sumamos los números de respuestas agrupando por sexos:
apply(table(Sexo,Respuestas,Pais), MARGIN=2, FUN=sum)
```

## 2.4 Gráficos de frecuencias

### 2.4.1 Diagrama de barras o bar plot

El tipo de gráfico más usado para representar variables cuantiativas o cualitativas son los **diagramas de barras (bar plots)**. Como su nombre indica, un diagrama de barras contiene, <u>para cada nivel de la variable, una barra de altura su frecuencia</u>.

Antes de continuar con los diagramas de barras, vamos a comentar un par de *tips* usados en el título de este gráfico. Por un lado, la marca `\n` dentro de una frase entre comillas introduce un cambio de línea. El mismo efecto se obtiene con un cambio de línea. Id con cuidado, porque ambos efectos se acumulan, así que si cambiais de línea después del \n, obtendréis una línea en blanco. Por otro lado, `\"` escribe unas comillas en el texto entre comillas.

**Ejemplo 12 <mark>[función barplot]</mark>** La manera más sencilla de **dibujar un diagrama de barras** de las frecuencias absolutas o relativas de una variable es usando la función `barplot` aplicada a la tabla correspondiente. Veamos dos ejemplos más de diagramas de barras, uno de frecuencias absolutas y uno de relativas:

```r
# Diagrama de barras de las frecuencias absolutas con la función barplot
x=c(3,2,5,1,3,1,5,6,2,2,2,1,3,5,2)
Respuestas=c("No","No","Sí","No","Sí","No","No","Sí")
barplot(table(x), main="Diagrama de barras de frecuencias absolutas de la variable \"x\"")
```

```r
# Diagrama de barras de las frecuencias relativas con la función barplot
barplot(prop.table(table(Respuestas)), 
        main="Diagrama de barras de frecuencias relativas\n de la variable \"Respuestas\"")
```

**<u>Descripción función barplot</u>.-** Como pasaba con `prop.table`, el argumento de `barplot` ha de ser una tabla, y, por consiguiente, se ha de aplicar al resultado de table o de prop.table, NUNCA al vector de datos original.

Habréis observado que en las funciones `barplot` anteriores hemos usado el parámetro `main` para poner título a los diagramas; en general, la función barplot admite los parámetros de plot que tienen sentido en el contexto de los diagramas de barras: `xlab, ylab, main, etc`. Los parámetros disponibles se pueden consultar en la ayuda de barplot. Aquí sólo vamos a comentar algunos.

**Ejemplo 13 <mark>[argumentos de la función barplot]</mark>** Se pueden especificar los colores de las barras usando el parámetro `col`. Si se iguala a un solo color, todas las barras serán de ese color, pero también se puede especificar un color para cada barra, igualando col a un vector de colores. En un diagrama con muchas barras, es conveniente usar un esquema adecuado de colores para ellas. Para ello se puede usar la librería $\tt RColorBrewer$.

```r
# Diagrama de barras de las frecuencias absolutas con el mismo color
barplot(table(Respuestas), col=c("green"))
```

```r
# Diagrama de barras de las frecuencias absolutas con distinto color
barplot(table(Respuestas), col=c("red","blue"))
```

```r
# Una opción interesante es dibujar las barras horizontales en vez de verticales: para hacerlo, se tiene que añadir 
# el parámetro horiz=TRUE
barplot(table(x), horiz=TRUE)
```

**Ejemplo 14 <mark>[función barplot con table caso bidimensional]</mark>** Si se aplica barplot a una tabla bidimensional, por defecto dibuja las barras de la segunda variable cortadas por la frecuencia de la primera variable: se le llama un diagrama de barras apiladas. En un diagrama de barras apiladas, las barras globales corresponden a los niveles de la variable que definen las columnas de la tabla, es decir, la segunda variable especificada dentro de table.

```r
# Obtenemos la tabla de frecuencias absolutas 
Respuestas=c("No","No","Sí","No","Sí","No","No","Sí")
Sexo=c("M","M","M","H","H","H","H","H")
table(Sexo,Respuestas)
```

```r
# Obtenemos un diagrama de barras apilado sobre una tabla bidimensional
barplot(table(Sexo,Respuestas))
```

**Ejemplo 15 <mark>[argumentos función barplot con table caso bidimensional]</mark>** En vez de organizar las barras de la primera variable en una sola barra vertical, se pueden dibujar una junto a la otra añadiendo el parámetro `beside=TRUE`. En este diagrama, cada bloque de barras representa un nivel de la variable de las columnas ($No$ y $Si$), y en cada uno de estos bloques las barras representan los niveles de las filas en su orden (en cada bloque, la barra de la izquierda corresponde a $H$ y la de la derecha a $M$).

```r
# Obteniéndose de esta manera un diagrama de barras por bloques
barplot(table(Sexo,Respuestas), beside=TRUE)
```

Los diagramas de barras tienen que mostrar la información de la manera más adecuada. Por ejemplo, si lo que nos interesa es la distribución de las respuestas por sexo, los bloques de barras tienen que corresponder a los sexos y las barras dentro de cada bloque a las respuestas. En este caso, convendría cambiar el orden de los vectores dentro de la table a la que aplicamos barplot, o trasponer la tabla antes de aplicarle barplot.

```r
# Veamos el caso transponiendo las variables
barplot(table(Respuestas,Sexo), beside=TRUE)
```

**Ejemplo 16 <mark>[leyenda barplot con table caso bidimensional]</mark>** Suele ser conveniente añadir a un diagrama de barras de dos variables una leyenda que indique qué nivel representa cada sector (en los diagramas de barras apiladas) o cada barra (en los diagramas de barras por bloques). Esto se puede realizar entrando el parámetro `legend.text` igualado a TRUE, si no queremos modificar los nombres de los niveles de las filas, o igualado a un vector con los nombres que les queremos asignar (en el orden que toque).

```r
# Añadimos una leyenda a nuestro gráfico para una mejor comprensión
barplot(table(Sexo,Respuestas), beside=TRUE, legend.text=TRUE)
```

Los extremos superiores de estas líneas dibujarán una curva: ésta es la densidad de la variable.

Las <u>medidas de tendencia central</u> son las que dan un valor representativo de todas las observaciones; las más importantes son:

+ La **moda**, que es el valor, o los valores, de máxima frecuencia (absoluta o relativa) se calcula como verermos más adelante.

+ La **mediana**, que representa el valor central en la lista ordenada de observaciones y se obtiene directamente a través de `median`.

# 4. ¿Cuáles son las principales medidas de posición? <a id="topic4"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Las <u>medidas de posición</u> estiman qué valores dividen la población en unas determinadas proporciones; los valores que determinan estas posiciones reciben el nombre de cuantiles. En este sentido, la mediana es también una medida de posición, puesto que divide la variable en dos mitades.

Dada una proporción $0<p<1$, el **cuantil** de orden $p$ de una variable cuantitativa, que denotaremos por $Q_p$, es el valor más pequeño tal que su frecuencia relativa acumulada es mayor o igual que $p$. En otras palabras, si tenemos un conjunto de datos $x_1,…,x_n$ y los ordenamos de menor a mayor, $Q_p$ es el número más pequeño que deja a su izquierda (incluyéndolo a él) como mínimo la fracción $p$ de los datos, es decir, $p\cdot n$ datos. De esta manera, la mediana viene a ser el cuantil $Q_{0.5}$. Con R, los cuantiles de orden $p$ de un vector $x$ se calculan con la función `quantile(x, p)`.

Algunos cuantiles tienen nombre propio:

+ La **mediana** es el cuantil $Q_0.5$.
+ Los **cuartiles** son los cuantiles $Q_{0.25}$, $Q_{0.5}$ y $Q_{0.75}$, y reciben, respectivamente, los nombres de primer cuartil, segundo cuartil (o mediana) y tercer cuartil. $Q_{0.25}$ será, pues, el menor valor que es mayor o igual que una cuarta parte de los datos, y $Q_{0.75}$, el menor valor que es mayor o igual que tres cuartas partes de los datos.
+ Los **deciles** son los cuantiles $Q_p$ con $p$ un múltiplo entero de 0.1: el primer decil es $Q_{0.1}$, el segundo decil es $Q_{0.2}$, y así sucesivamente.
+ Los **percentiles** son los cuantiles $Q_p$ con $p$ un múltiplo entero de 0.01.

**Ejemplo 22 <mark>[función quantile]</mark>** Consideremos un experimento de lanzamientos de un dado. Lo lanzamos $30$ veces y obtenemos los resultados siguientes: $2$, $4$, $5$, $6$, $3$, $2$, $4$, $5$, $1$, $2$, $1$, $3$, $4$, $2$, $3$, $4$, $1$, $2$, $5$, $6$, $5$, $5$, $3$, $2$, $1$, $3$, $4$, $2$, $2$, $1$.

```r
# Ordenamos estos resultados de menor a mayor
dados=c(2,4,5,6,3,2,4,5,1,2,1,3,4,2,3,4,1,2,5,6,5,5,3,2,1,3,4,2,2,1)
dados=sort(dados)
dados
```

Si nos pidieran el cuantil $Q_{0.2}$, sería el primer elemento en esta lista ordenada que fuera mayor o igual que, como mínimo, el 20% de los datos. Como el $20\%$ de 30 es 6, sería el sexto elemento.

```r
# Obtenemos el cuantil 20
quantile(dados, 0.2); dados[6]
```

Si nos pidieran en cambio el cuantil $Q_{0.65}$, sería el primer elemento en esta lista ordenada mayor o igual que, como mínimo, el 65% de los datos. Como el 65% de 30 es 19.5, sería el vigésimo elemento.

```r
# Obtenemos el cuantil 65
quantile(dados, 0.65); dados[20]
```

# 5. ¿Cuáles son las principales medidas de dispersión? <a id="topic5"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Las <u>medidas de dispersión</u> evalúan lo disperso que están los datos. Las más importantes son:

+ El **rango**, o recorrido, que es la diferencia entre el máximo y el mínimo de las observaciones. La función `range(x)` nos da sus valores mínimo y máximo. Por lo tanto, su rango se puede calcular con `diff(range(x))`.


+ El **rango intercuartílico**, que es la diferencia $Q_{0.75}−Q_{0.25}$. También se llama a veces rango intercuartílico al intervalo intercuartílico: el intervalo $\left[ {Q_{0.25},Q_{0.75}} \right]$. El rango intercuartílico se calcula con `IQR(x)`. 


+ La **varianza**, que es la media aritmética de las diferencias al cuadrado entre los datos $x_i$ y la media $\bar x$
de la variable y la denotamos por $s^2$. L varianza se obtiene con la función `var()`.


+ La **desviación típica**, que es la raíz cuadrada positiva $s$ de la varianza: $s=\sqrt s^2$. La desviación típica se calcula con la función `sd()`.

**Ejemplo 23 <mark>[funciones medidas dispersión]</mark>** Vamos a definir un vector de números al azar para poder obtener cada una de las medidas de dispersión enumeradas.

```r
# Consideremos el vector
x=c(1,2,3,4,5,6,2,3,2,3,4,2,2,3,2,2,5,7,3,4,2,1,3,6)

# Su rango es
diff(range(x))
```

```r
# Su rango intercuartílico es
IQR(x)
```

```r
# Su varianza es
var(x)
```

```r
# Su desviación típica es
sd(x)
```
