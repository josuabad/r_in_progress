## 2.1 Tablas de frecuencias unidimensionales

Una <u>tabla de frecuencias</u> muestra de forma ordenada un conjunto de datos estadsticos y a cada uno de ellos le asigna una frecuencia que, en pocas palabras, son las **veces que se repite un nmero o dato**. Puedes usar las tablas de frecuencias para ordenar variables cuantitativas o cualitativas. Existen diferentes tipos de tablas de frecuencias:

1. **Frecuencias absolutas**: son el nmero de veces que se repite un nmero en un conjunto de datos.
2. **Frecuencias relativas**: corresponden a las veces que se repite un nmero en un conjunto de datos respecto al total y se expresa en porcentajes (%).

**Ejemplo 1 <mark>[frecuencia absoluta]</mark>** Supongamos que tenemos una variable cuantitativa o cualitativa guardada en un vector o factor, veamos cmo calculamos la frecuencia absoluta y su interpretacin. 

```r
# Generamos el vector de nmeros
x=c(3,2,5,1,3,1,5,6,2,2,2,1,3,5,2)
x
```

```r
# Generamos el factor con las respuestas Si/No
Respuestas=factor(c("No","No","S","No","S","No","No","S"))
Respuestas
```

La **tabla de frecuencias absolutas** de un vector o factor que representa una variable cuantitativa o cualitativa se calcula con la funcin `table`. 

```r
# Obtenemos la tabla de frecuancias absolutas del vector de nmeros
table(x)
```

```r
# Obtenemos la tabla de frecuancias absolutas del factor de respuestas
table(Respuestas)
```

**<u>Objeto creado por la funcin table</u>.-** El resultado de una funcin table es un objeto de datos de un tipo nuevo: una *tabla de contingencia*. Como vemos, al aplicar table a un vector obtenemos una tabla unidimensional formada por una fila con los niveles de la variable y una segunda fila donde, debajo de cada nivel, aparece su frecuencia absoluta en el vector.

```r
# Los nombres de las columnas de una tabla unidimensional se obtienen con la funcin names()
names(table(x))
```

```r
# Los nombres de las columnas de una tabla unidimensional se obtienen con la funcin names()
names(table(Respuestas))
```

**<u>Nombres de niveles ocultos</u>.-** Habris observado que tras aplicar la funcin table a un vector slo aparecen los nombres de los niveles presentes en el vector. Si el tipo de datos cualitativos usado tena ms **niveles** y queremos que **aparezcan explcitamente** en la tabla (con frecuencia $0$), hay que *transformar* el *vector* en un *factor* con los niveles deseados.

```r
# Creamos un nuevo factor y le asignamos los niveles que queremos tener en cuenta
z=factor(x, levels=1:7)  #Los niveles sern 1,2,3,4,5,6,7
z
```

```r
# Obtenemos la tabla de frecuancias absolutas del factor de nmeros creado
table(z)
```

**Ejemplo 2 <mark>[acceder a elementos de una table]</mark>** A efectos prcticos, podemos pensar que una tabla unidimensional es como un vector de nmeros donde cada entrada est identificada por un nombre: el de su columna. Para referirnos a una entrada de una tabla unidimensional, podemos usar tanto su posicin como su nombre (entre comillas, aunque sea un nmero). Veamos algunos ejemplos. 

```r
# La cuarta columna de table(z)
table(z)[4]
# La columna de table(z) correspondiente al nivel 5:
table(z)["5"]
# El triple de la segunda columna de table(z)
3*table(z)[2]
```

**Ejemplo 3 <mark>[frecuencia relativa]</mark>** La *tabla de frecuencias relativas* de un vector se puede calcular aplicando la funcin `prop.table` a su table. El resultado vuelve a ser una tabla de contingencia unidimensional.

```r
# Obtenemos la tabla de frecuancias relativas del vector de nmeros
prop.table(table(x))
```

```r
# Obtenemos la tabla de frecuancias relativas del factor de respuestas
prop.table(table(Respuestas))
```

**<u>Importante</u>.-** Tener en cuenta que la funcin prop.table se tiene que **aplicar al resultado de table**, NO al vector original. 

Si aplicamos prop.table a un vector de palabras o a un factor, **dar un error**, pero si la aplicamos a un vector de nmeros, nos dar una tabla. Esta tabla NO es la tabla de frecuencias relativas de la variable cualitativa representada por el vector, sino la de una variable que tuviera como tabla de frecuencias absolutas este vector de nmeros, entendiendo que cada entrada del vector representa la frecuencia de un nivel diferente. 

```r
# Aplicacin errnea de la funcin prop.table() que no es la tabla de frecuencias relativas de x
prop.table(x)
```

**Ejemplo 4 <mark>[frecuencia relativa utilizando table]</mark>** Tambin podemos calcular la tabla de frecuencias relativas de un vector dividiendo el resultado de table por el nmero de observaciones del siguiente modo.

```r
# Obtenemos la tabla de frecuancias relativas del vector de nmeros operando con la funcin table()
table(x)/length(x)
table(x)/sum(table(x)) # Otra alternativa
```

**Ejemplo 5 <mark>[moda utilizando frecuencia absoluta]</mark>** Dados un vector x y un nmero natural n, la instruccin `names(which(table(x)==n)` nos da los niveles de x que tienen frecuencia absoluta igual a *n*.

```r
# Obtenemos la frecuencia absoluta del vector de nmeros que sea igual a n operando con la funcin table() y which()
table(x)
names(which(table(x)==3)) # cuando n = 3
```

```r
# Obtenemos la moda del vector de nmeros operando con la funcin table() y which()
names(which(table(x)==max(table(x))))
names(which(table(Respuestas)==max(table(Respuestas))))
```

## 2.2 Tablas de frecuencias bidimensionales

La funcin table tambin permite construir **tablas de frecuencias conjuntas de dos o ms variables**. 

**Ejemplo 6 <mark>[frecuencia absoluta para 2 variables]</mark>** Supongamos que el vector `Respuestas`, de la seccin anterior, contiene las respuestas a una pregunta dadas por unos individuos cuyos sexos tenemos almacenados en un vector `Sexo`, en el mismo orden que sus respuestas. En este caso, podemos construir una tabla que nos diga cuntos individuos de cada sexo han dado cada respuesta.

```r
# Obtenemos la tabla de frecuancias absolutas de las dos variables Respuestas y Sexo
Respuestas=c("No","No","S","No","S","No","No","S")
Sexo=c("M","M","M","H","H","H","H","H") # H es hombre, M es mujer
table(Respuestas,Sexo)
```

```r
# Modificamos el orden de las filas y columnas
table(Sexo,Respuestas)
```

**<u>Transponer filas y columnas</u>.-** Como podemos ver, en una tabla bidimensional producida aplicando table a dos vectores, los niveles del primer vector en el argumento definen las filas, y los del segundo, las columnas. As, en table(Respuestas,Sexo), las filas corresponden a las respuestas y las columnas a los sexos. Para intercambiar filas por columnas, es decir, para trasponer la tabla sin tener que recalcularla, podemos usar la misma funcin `t()` que usamos para trasponer matrices:

```r
# Transponemos el orden de las filas y columnas 
t(table(Respuestas,Sexo))
```

**Ejemplo 7 <mark>[acceder a elementos de una table bidimensional]</mark>** Para referirnos a una entrada de una tabla bidimensional podemos usar corchetes como si estuviramos en una matriz o un data frame. Dentro de los corchetes, tanto podemos usar los ndices como los nombres (entre comillas) de los niveles.

```r
# El elemento de la 1 fila y la 2 columna de table(Respuestas,Sexo)
table(Respuestas,Sexo)[1,2]

# El elemento de la fila con valor igual a 'No' y la columna 'M' de table(Respuestas,Sexo)
table(Respuestas,Sexo)["No","M"]
```

**Ejemplo 8 <mark>[frecuencias relativas bidimensionales: globales vs marginales]</mark>** Como en el caso unidimensional, la funcin `prop.table` sirve para calcular tablas bidimensionales de frecuencias relativas conjuntas de pares de variables. Pero en el caso bidimensional tenemos dos tipos de frecuencias relativas, que definen, para cada par de variables, tres tablas diferentes:

+ Las frecuencias relativas **globales**: para cada par de niveles, uno de cada variable, la <u>fraccin de individuos que pertenecen a ambos niveles respecto del total de la muestra</u>; por ejemplo, la fraccin de mujeres que han contestado que s respecto del total de la muestra sera la frecuencia relativa global del par (mujer, s). La tabla de frecuencias relativas globales se calcula aplicando sin ms la funcin `prop.table` a la table.


+ Las frecuencias relativas **marginales**: dentro de cada nivel de una variable, y para cada nivel de la otra, la <u>fraccin de individuos que pertenecen al segundo nivel respecto del total de la subpoblacin definida por el primer nivel</u>; por ejemplo, la fraccin de mujeres que han contestado que s respecto del total de mujeres sera una frecuencia relativa marginal. Dadas dos variables, <u>se pueden calcular dos familias de frecuencias relativas marginales</u>, segn cul sea la variable que defina las subpoblaciones en las que calculemos las frecuencias relativas de los niveles de la otra variable. La variable que define las subpoblaciones en las que las calculamos se indica con el parmetro `margin`. Es decir, con `margin=1` especificamos que la variable que define las subpoblaciones es la primera, y que, por lo tanto, las frecuencias relativas se calculan dentro de las filas; en cambio, con `margin=2` especificamos que la variable que define las subpoblaciones es la segunda, por lo que las frecuencias relativas se calculan por columnas; `margin=NULL` es su valor por defecto y, por lo tanto, equivalente a no especificar este parmetro.

**<u>Interpretacin resultados</u>.-** De esta manera, por ejemplo, en la primera tabla podemos leer que un 25% del total de la muestra son mujeres que han contestado que no; en la segunda tabla, que un 66.67% de las mujeres han contestado que no; y en la tercera tabla, que las mujeres representan el 40% del total de las personas que han contestado que no.

```r
# La tabla prop.table(table(Sexo,Respuestas)) nos da la fraccin del total que representa 
# cada pareja (sexo, respuesta)
prop.table(table(Sexo,Respuestas))
```

```r
# La tabla prop.table(table(Sexo,Respuestas), margin=1) nos da la fraccin que representa 
# cada respuesta dentro de cada sexo
prop.table(table(Sexo,Respuestas), margin=1)
```

```r
# La tabla prop.table(table(Sexo,Respuestas), margin=2) nos da la fraccin que representa 
# cada sexo dentro de cada respuesta
prop.table(table(Sexo,Respuestas), margin=2)
```

**Ejemplo 9 <mark>[todas las frecuencias relativas en una tabla]</mark>** La funcin `CrossTable` de la librera $\tt gmodels$ permite producir (especificando el parmetro `prop.chisq=FALSE`) un resumen de la tabla de frecuencias absolutas y las tres tablas de frecuencias relativas de dos variables en un formato adecuado para su visualizacin.

```r
# Instalamos la librera gmodels
install.packages("gmodels")
# Cargamos la librera gmodels
library(gmodels) 
# Obtenemos los resultados de todas las frecuencias relativas en una tabla
CrossTable(Sexo, Respuestas, prop.chisq=FALSE)
```

**<u>Descripcin funcin CrossTable</u>.-** La leyenda Cell Contents explica los contenidos de cada celda de la tabla: en este caso, y en orden descendente, la frecuencia absoluta N, la frecuencia relativa por filas, la frecuencia relativa por columnas, y la frecuencia relativa global. Asimismo, se muestran las celdas de los mrgenes, con las frecuencias absolutas y relativas de cada fila (en la columna Row Total) y cada columna (en la fila Column Total). Esta funcin dispone de muchos parmetros que permiten modificar el contenido de las celdas y que podis consultar en su Ayuda.

```r
# Ayuda de la funcin CrossTable
help(CrossTable)
```

## 2.3 Tablas de frecuencias multidimensionales

En general, podemos calcular **tablas de frecuencias de cualquier nmero de variables**, no slo de una o dos. El manejo de estas tablas multidimensionales es similar al caso bidimensional, simplemente recordando que ahora hay ms variables que tener en cuenta en el momento, por ejemplo, de especificar entradas o de calcular frecuencias relativas marginales. 

**Ejemplo 10 <mark>[frecuencia absoluta para 3 variables]</mark>** Supongamos que, adems de los vectores utilizados de `Respuestas`y `Sexo`, tenemos un tercer vector con las nacionalidades de los individuos `Pais`.

```r
# Leemos las variables que vamos a utilizar
Respuestas=c("No","No","S","No","S","No","No","S")
Sexo=c("M","M","M","H","H","H","H","H")
Pais=c("Francia","Alemania","Italia","Italia","Italia","Italia","Alemania","Francia")
```

```r
# Podemos calcular entonces una tabla de frecuencias absolutas para las ternas (sexo, respuesta, pas).
table(Sexo,Respuestas,Pais)
```

**<u>Descripcin funcin table caso multidimensional</u>.-** R muestra la tabla tridimensional que obtenemos como una lista de tablas bidimensionales `table(Sexo,Respuestas)`, separando la poblacin segn los niveles de la tercera variable. Si no os gusta esta manera de visualizar una tabla tridimensional, una alternativa es usar la funcin `ftable`, que la mostrar en lo que se llama formato plano:

```r
# Visualizacin compacta de una tabla tridimensional
ftable(Sexo,Respuestas,Pais)
```

```r
# Los parmetros row.vars y col.vars de ftable permiten especificar qu variables queremos que aparezcan 
# como filas o como columnas, respectivamente
ftable(Sexo,Respuestas,Pais, col.vars=c("Sexo","Respuestas"))
```

```r
# Para referirnos a una entrada, o a una subtabla, de una tabla podemos usar corchetes
table(Sexo,Respuestas,Pais)["H","S","Italia"] # Hombre, Si, Italianos
```

```r
table(Sexo,Respuestas,Pais)[ , ,"Italia"] # Italianos
```

```r
table(Sexo,Respuestas,Pais)[  ,"S","Italia"] # Si, Italianos
```

```r
table(Sexo,Respuestas,Pais)["M", ,"Italia"] # Mujer, Italianos
```

**Ejemplo 10 <mark>[frecuencias relativas para 3 variables]</mark>** En una tabla multidimensional, podemos calcular frecuencias relativas marginales respecto de los niveles de una variable o respecto de combinaciones de niveles de varias variables. Por ejemplo, <u>las frecuencias relativas marginales de las respuestas en cada combinacin (sexo, pas)</u>. 

Como en el caso bidimensional, las tablas correspondientes se calculan aplicando `prop.table` a la tabla de frecuencias absolutas, y especificando con el parmetro `margin` las dimensiones, o combinaciones de dimensiones, respecto de las que calculamos las frecuencias relativas. Si no se especifica el parmetro margin, se obtiene la tabla de frecuencias relativas globales.

```r
# La tabla prop.table(table(Sexo,Respuestas,Pais)) nos da la fraccin que representa 
# cada terna (sexo, respuesta, pas) dentro del total de la muestra:
prop.table(table(Sexo,Respuestas,Pais))
```

```r
# La tabla prop.table(table(Sexo,Respuestas,Pais), margin=3) nos da la fraccin que representa 
# cada pareja (sexo, respuesta) dentro de cada pas:
prop.table(table(Sexo,Respuestas,Pais), margin=3)
```

**<u>Interpretacin resultados</u>.-** En la primera tabla podemos leer que los hombres italianos que han contestado afirmativamente forman un 12.5% del total de individuos; en la segunda, que los hombres que han contestado que s representan un 25% del total de individuos italianos; y en la tercera, que un 33.33% del total de hombres espaoles ha contestado que s.

```r
# Al aplicar prop.table al resultado de una ftable, la tabla resultante sigue en formato plano:
prop.table(ftable(Sexo,Respuestas,Pais)) 
```

**Ejemplo 11 <mark>[apply con la funcin table]</mark>** Podemos aplicar funciones a filas o columnas de una tabla multidimensional con la funcin `apply`, especificando en el parmetro $\tt FUN$ la funcin y en el parmetro $\tt  MARGIN$ la variable o combinacin de variables a la que aplicamos la funcin (son las variables que aparecern en la tabla resultante). Observar las dos instrucciones siguientes y sus resultados:

```r
# Sumamos los nmeros de respuestas agrupando por combinaciones (sexo,respuesta):
apply(table(Sexo,Respuestas,Pais), MARGIN=c(1,2), FUN=sum)
```

```r
# Sumamos los nmeros de respuestas agrupando por sexos:
apply(table(Sexo,Respuestas,Pais), MARGIN=2, FUN=sum)
```

## 2.4 Grficos de frecuencias

### 2.4.1 Diagrama de barras o bar plot

El tipo de grfico ms usado para representar variables cuantiativas o cualitativas son los **diagramas de barras (bar plots)**. Como su nombre indica, un diagrama de barras contiene, <u>para cada nivel de la variable, una barra de altura su frecuencia</u>.

Antes de continuar con los diagramas de barras, vamos a comentar un par de *tips* usados en el ttulo de este grfico. Por un lado, la marca `\n` dentro de una frase entre comillas introduce un cambio de lnea. El mismo efecto se obtiene con un cambio de lnea. Id con cuidado, porque ambos efectos se acumulan, as que si cambiais de lnea despus del \n, obtendris una lnea en blanco. Por otro lado, `\"` escribe unas comillas en el texto entre comillas.

**Ejemplo 12 <mark>[funcin barplot]</mark>** La manera ms sencilla de **dibujar un diagrama de barras** de las frecuencias absolutas o relativas de una variable es usando la funcin `barplot` aplicada a la tabla correspondiente. Veamos dos ejemplos ms de diagramas de barras, uno de frecuencias absolutas y uno de relativas:

```r
# Diagrama de barras de las frecuencias absolutas con la funcin barplot
x=c(3,2,5,1,3,1,5,6,2,2,2,1,3,5,2)
Respuestas=c("No","No","S","No","S","No","No","S")
barplot(table(x), main="Diagrama de barras de frecuencias absolutas de la variable \"x\"")
```

```r
# Diagrama de barras de las frecuencias relativas con la funcin barplot
barplot(prop.table(table(Respuestas)), 
        main="Diagrama de barras de frecuencias relativas\n de la variable \"Respuestas\"")
```

**<u>Descripcin funcin barplot</u>.-** Como pasaba con `prop.table`, el argumento de `barplot` ha de ser una tabla, y, por consiguiente, se ha de aplicar al resultado de table o de prop.table, NUNCA al vector de datos original.

Habris observado que en las funciones `barplot` anteriores hemos usado el parmetro `main` para poner ttulo a los diagramas; en general, la funcin barplot admite los parmetros de plot que tienen sentido en el contexto de los diagramas de barras: `xlab, ylab, main, etc`. Los parmetros disponibles se pueden consultar en la ayuda de barplot. Aqu slo vamos a comentar algunos.

**Ejemplo 13 <mark>[argumentos de la funcin barplot]</mark>** Se pueden especificar los colores de las barras usando el parmetro `col`. Si se iguala a un solo color, todas las barras sern de ese color, pero tambin se puede especificar un color para cada barra, igualando col a un vector de colores. En un diagrama con muchas barras, es conveniente usar un esquema adecuado de colores para ellas. Para ello se puede usar la librera $\tt RColorBrewer$.

```r
# Diagrama de barras de las frecuencias absolutas con el mismo color
barplot(table(Respuestas), col=c("green"))
```

```r
# Diagrama de barras de las frecuencias absolutas con distinto color
barplot(table(Respuestas), col=c("red","blue"))
```

```r
# Una opcin interesante es dibujar las barras horizontales en vez de verticales: para hacerlo, se tiene que aadir 
# el parmetro horiz=TRUE
barplot(table(x), horiz=TRUE)
```

**Ejemplo 14 <mark>[funcin barplot con table caso bidimensional]</mark>** Si se aplica barplot a una tabla bidimensional, por defecto dibuja las barras de la segunda variable cortadas por la frecuencia de la primera variable: se le llama un diagrama de barras apiladas. En un diagrama de barras apiladas, las barras globales corresponden a los niveles de la variable que definen las columnas de la tabla, es decir, la segunda variable especificada dentro de table.

```r
# Obtenemos la tabla de frecuencias absolutas 
Respuestas=c("No","No","S","No","S","No","No","S")
Sexo=c("M","M","M","H","H","H","H","H")
table(Sexo,Respuestas)
```

```r
# Obtenemos un diagrama de barras apilado sobre una tabla bidimensional
barplot(table(Sexo,Respuestas))
```

**Ejemplo 15 <mark>[argumentos funcin barplot con table caso bidimensional]</mark>** En vez de organizar las barras de la primera variable en una sola barra vertical, se pueden dibujar una junto a la otra aadiendo el parmetro `beside=TRUE`. En este diagrama, cada bloque de barras representa un nivel de la variable de las columnas ($No$ y $Si$), y en cada uno de estos bloques las barras representan los niveles de las filas en su orden (en cada bloque, la barra de la izquierda corresponde a $H$ y la de la derecha a $M$).

```r
# Obtenindose de esta manera un diagrama de barras por bloques
barplot(table(Sexo,Respuestas), beside=TRUE)
```

Los diagramas de barras tienen que mostrar la informacin de la manera ms adecuada. Por ejemplo, si lo que nos interesa es la distribucin de las respuestas por sexo, los bloques de barras tienen que corresponder a los sexos y las barras dentro de cada bloque a las respuestas. En este caso, convendra cambiar el orden de los vectores dentro de la table a la que aplicamos barplot, o trasponer la tabla antes de aplicarle barplot.

```r
# Veamos el caso transponiendo las variables
barplot(table(Respuestas,Sexo), beside=TRUE)
```

**Ejemplo 16 <mark>[leyenda barplot con table caso bidimensional]</mark>** Suele ser conveniente aadir a un diagrama de barras de dos variables una leyenda que indique qu nivel representa cada sector (en los diagramas de barras apiladas) o cada barra (en los diagramas de barras por bloques). Esto se puede realizar entrando el parmetro `legend.text` igualado a TRUE, si no queremos modificar los nombres de los niveles de las filas, o igualado a un vector con los nombres que les queremos asignar (en el orden que toque).

```r
# Aadimos una leyenda a nuestro grfico para una mejor comprensin
barplot(table(Sexo,Respuestas), beside=TRUE, legend.text=TRUE)
```

```r
# Otro ejemplo utilizando todos los argumentos de la funcin barplot vistos
barplot(table(Respuestas,Sexo), beside=TRUE, names=c("Men","Women"),
        col=c("red","blue"), legend.text=c("No","Yes"))
```

### 2.4.2 Diagrama de sectores o pie chart

Un tipo muy popular de representacin grfica son los **diagramas circulares**. # 4. Cules son las principales medidas de posicin? <a id="topic4"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Las <u>medidas de posicin</u> estiman qu valores dividen la poblacin en unas determinadas proporciones; los valores que determinan estas posiciones reciben el nombre de cuantiles. En este sentido, la mediana es tambin una medida de posicin, puesto que divide la variable en dos mitades.

Dada una proporcin $0<p<1$, el **cuantil** de orden $p$ de una variable cuantitativa, que denotaremos por $Q_p$, es el valor ms pequeo tal que su frecuencia relativa acumulada es mayor o igual que $p$. En otras palabras, si tenemos un conjunto de datos $x_1,,x_n$ y los ordenamos de menor a mayor, $Q_p$ es el nmero ms pequeo que deja a su izquierda (incluyndolo a l) como mnimo la fraccin $p$ de los datos, es decir, $p\cdot n$ datos. De esta manera, la mediana viene a ser el cuantil $Q_{0.5}$. Con R, los cuantiles de orden $p$ de un vector $x$ se calculan con la funcin `quantile(x, p)`.

Algunos cuantiles tienen nombre propio:

+ La **mediana** es el cuantil $Q_0.5$.
+ Los **cuartiles** son los cuantiles $Q_{0.25}$, $Q_{0.5}$ y $Q_{0.75}$, y reciben, respectivamente, los nombres de primer cuartil, segundo cuartil (o mediana) y tercer cuartil. $Q_{0.25}$ ser, pues, el menor valor que es mayor o igual que una cuarta parte de los datos, y $Q_{0.75}$, el menor valor que es mayor o igual que tres cuartas partes de los datos.
+ Los **deciles** son los cuantiles $Q_p$ con $p$ un mltiplo entero de 0.1: el primer decil es $Q_{0.1}$, el segundo decil es $Q_{0.2}$, y as sucesivamente.
+ Los **percentiles** son los cuantiles $Q_p$ con $p$ un mltiplo entero de 0.01.

**Ejemplo 22 <mark>[funcin quantile]</mark>** Consideremos un experimento de lanzamientos de un dado. Lo lanzamos $30$ veces y obtenemos los resultados siguientes: $2$, $4$, $5$, $6$, $3$, $2$, $4$, $5$, $1$, $2$, $1$, $3$, $4$, $2$, $3$, $4$, $1$, $2$, $5$, $6$, $5$, $5$, $3$, $2$, $1$, $3$, $4$, $2$, $2$, $1$.

```r
# Ordenamos estos resultados de menor a mayor
dados=c(2,4,5,6,3,2,4,5,1,2,1,3,4,2,3,4,1,2,5,6,5,5,3,2,1,3,4,2,2,1)
dados=sort(dados)
dados
```

Si nos pidieran el cuantil $Q_{0.2}$, sera el primer elemento en esta lista ordenada que fuera mayor o igual que, como mnimo, el 20% de los datos. Como el $20\%$ de 30 es 6, sera el sexto elemento.

```r
# Obtenemos el cuantil 20
quantile(dados, 0.2); dados[6]
```

Si nos pidieran en cambio el cuantil $Q_{0.65}$, sera el primer elemento en esta lista ordenada mayor o igual que, como mnimo, el 65% de los datos. Como el 65% de 30 es 19.5, sera el vigsimo elemento.

```r
# Obtenemos el cuantil 65
quantile(dados, 0.65); dados[20]
```

# 5. Cules son las principales medidas de dispersin? <a id="topic5"></a>&nbsp;&nbsp;<a href="#index"><i class="fa fa-list-alt" aria-hidden="true"></i></a>


Las <u>medidas de dispersin</u> evalan lo disperso que estn los datos. Las ms importantes son:

+ El **rango**, o recorrido, que es la diferencia entre el mximo y el mnimo de las observaciones. La funcin `range(x)` nos da sus valores mnimo y mximo. Por lo tanto, su rango se puede calcular con `diff(range(x))`.


+ El **rango intercuartlico**, que es la diferencia $Q_{0.75}Q_{0.25}$. Tambin se llama a veces rango intercuartlico al intervalo intercuartlico: el intervalo $\left[ {Q_{0.25},Q_{0.75}} \right]$. El rango intercuartlico se calcula con `IQR(x)`. 


+ La **varianza**, que es la media aritmtica de las diferencias al cuadrado entre los datos $x_i$ y la media $\bar x$
de la variable y la denotamos por $s^2$. L varianza se obtiene con la funcin `var()`.


+ La **desviacin tpica**, que es la raz cuadrada positiva $s$ de la varianza: $s=\sqrt s^2$. La desviacin tpica se calcula con la funcin `sd()`.

**Ejemplo 23 <mark>[funciones medidas dispersin]</mark>** Vamos a definir un vector de nmeros al azar para poder obtener cada una de las medidas de dispersin enumeradas.

```r
# Consideremos el vector
x=c(1,2,3,4,5,6,2,3,2,3,4,2,2,3,2,2,5,7,3,4,2,1,3,6)

# Su rango es
diff(range(x))
```

```r
# Su rango intercuartlico es
IQR(x)
```

```r
# Su varianza es
var(x)
```

```r
# Su desviacin tpica es
sd(x)
```
